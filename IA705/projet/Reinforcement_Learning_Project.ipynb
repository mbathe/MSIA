````python
# --------------------------------------------------------
# Example Code: Using PyBullet for Robotic Arm Simulation
# --------------------------------------------------------
# This example demonstrates how to:
# - Set up a PyBullet environment
# - Load a robotic arm (KUKA iiwa) with a gripper
# - Load objects (table, cube) for interaction
# - Attach the gripper to the robotic arm
# - Control the robotic arm using inverse kinematics
# - Move the robotic arm to pick and place an object
# - Capture frames from the simulation and save as a video
#
# This code provides an **overview** of how PyBullet can be used for robotic simulations.
# For a **deeper understanding**, please refer to the **tutorials and resources** provided
# in the documentation section of this project.
# --------------------------------------------------------

# Import necessary libraries for physics simulation, image processing, and visualization
import os
import time
import math
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import pybullet as p
import pybullet_data
import cv2
import imageio_ffmpeg
from base64 import b64encode
from IPython.display import HTML

# --------------------------------------------------------
# SETUP: Initialize PyBullet simulation
# --------------------------------------------------------

# Use p.GUI for a graphical interface or p.DIRECT for headless mode (no GUI)
p.connect(p.DIRECT)

# Add PyBullet's built-in data directory for loading objects
p.setAdditionalSearchPath(pybullet_data.getDataPath())

# Set gravity in the environment (0, 0, -10) simulates Earth's gravity (negative z direction)
p.setGravity(0, 0, -10)


# --------------------------------------------------------
# LOAD OBJECTS INTO SIMULATION
# --------------------------------------------------------

# Load a static ground plane
plane_id = p.loadURDF("plane.urdf")

# Load a KUKA robotic arm at a specific position
kuka_id = p.loadURDF("kuka_iiwa/model_vr_limits.urdf",
                      1.4, -0.2, 0.6,  # Base position (x, y, z)
                      0, 0, 0, 1)      # Base orientation (quaternion)

# Load a gripper and attach it to the robotic arm
kuka_gripper_id = p.loadSDF("gripper/wsg50_one_motor_gripper_new_free_base.sdf")[0]

# Load a table in the environment
table_id = p.loadURDF("table/table.urdf",
                      basePosition=[1.0, -0.2, 0.0],
                      baseOrientation=[0, 0, 0.7071, 0.7071])

# Load a cube (object to interact with) at a specific position
cube_id = p.loadURDF("cube.urdf", basePosition=[0.85, -0.2, 0.65], globalScaling=0.05)

# --------------------------------------------------------
# ATTACH GRIPPER TO ROBOTIC ARM
# --------------------------------------------------------

# Create a fixed constraint between the arm and the gripper
kuka_cid = p.createConstraint(kuka_id, 6, kuka_gripper_id, 0,
                              p.JOINT_FIXED, [0, 0, 0], [0, 0, 0.05], [0, 0, 0])

# Add a gear constraint to control the gripper fingers
kuka_cid2 = p.createConstraint(kuka_gripper_id, 4, kuka_gripper_id, 6,
                               jointType=p.JOINT_GEAR, jointAxis=[1,1,1],
                               parentFramePosition=[0,0,0], childFramePosition=[0,0,0])

# Adjust constraint properties to control the gripper
p.changeConstraint(kuka_cid2, gearRatio=-1, erp=0.5, relativePositionTarget=0, maxForce=100)

# --------------------------------------------------------
# RESET ROBOTIC ARM TO INITIAL POSITION
# --------------------------------------------------------

# Define initial joint positions for the robotic arm
jointPositions = [-0.0, -0.0, 0.0, 1.570793, 0.0, -1.036725, 0.000001]

# Reset each joint to the predefined position
for jointIndex in range(p.getNumJoints(kuka_id)):
    p.resetJointState(kuka_id, jointIndex, jointPositions[jointIndex])
    p.setJointMotorControl2(kuka_id, jointIndex, p.POSITION_CONTROL,
                            jointPositions[jointIndex], 0)

# Reset the gripper's position
p.resetBasePositionAndOrientation(kuka_gripper_id, [0.923103, -0.200000, 1.250036],
                                  [-0.000000, 0.964531, -0.000002, -0.263970])

# Define gripper joint positions and reset them
jointPositions = [0.000000, -0.011130, -0.206421, 0.205143, -0.009999, 0.000000, -0.010055, 0.000000]
for jointIndex in range(p.getNumJoints(kuka_gripper_id)):
    p.resetJointState(kuka_gripper_id, jointIndex, jointPositions[jointIndex])
    p.setJointMotorControl2(kuka_gripper_id, jointIndex, p.POSITION_CONTROL,
                            jointPositions[jointIndex], 0)

# Define key parameters
num_joints = p.getNumJoints(kuka_id)  # Number of joints in the KUKA arm
kuka_end_effector_idx = 6  # Define the end-effector index

# --------------------------------------------------------
# CAMERA SETUP FOR VIDEO RECORDING
# --------------------------------------------------------

# Define camera parameters
cam_target_pos = [.95, -0.2, 0.2]  # Camera focus point
cam_distance = 2.05  # Distance from the scene
cam_yaw, cam_pitch, cam_roll = -50, -40, 0  # Camera angles
cam_width, cam_height = 480, 360  # Image resolution

# Initialize video writer
vid = imageio_ffmpeg.write_frames('vid.mp4', (cam_width, cam_height), fps=30)
vid.send(None)  # Start video recording

# --------------------------------------------------------
# SIMULATION LOOP (MOVE ROBOT ARM)
# --------------------------------------------------------

for t in range(750):
    print(f'\rTimestep {t}...', end='')

    # Capture and record frames every 8 steps (to maintain 30 FPS video)
    if t % 8 == 0:
        cam_view_matrix = p.computeViewMatrixFromYawPitchRoll(cam_target_pos, cam_distance, cam_yaw, cam_pitch, cam_roll, 2)
        cam_projection_matrix = p.computeProjectionMatrixFOV(60, cam_width*1./cam_height, 0.01, 100)
        image = p.getCameraImage(cam_width, cam_height, cam_view_matrix, cam_projection_matrix)[2][:, :, :3]
        vid.send(np.ascontiguousarray(image))  # Save frame to video

    # Define movement phases for the robotic arm and gripper
    target_pos, gripper_val = [0.85, -0.2, 0.97], 0
    if 150 <= t < 250:
        target_pos, gripper_val = [0.85, -0.2, 0.97], 1  # Grab object
    elif 250 <= t < 400:
        target_pos, gripper_val = [0.85, -0.2, 0.97 + 0.13*(t-250)/150.], 1  # Lift object
    elif 400 <= t < 600:
        target_pos, gripper_val = [0.85, -0.2 + 0.4*(t-400)/200., 1.1], 1  # Move object sideways
    elif 600 <= t < 700:
        target_pos, gripper_val = [0.85, 0.2, 1.1], 1  # Hold object
    elif t >= 700:
        target_pos, gripper_val = [0.85, 0.2, 1.1], 0  # Release object

    # Compute inverse kinematics to move the robotic arm
    target_orn = p.getQuaternionFromEuler([0, 1.01*math.pi, 0])
    joint_poses = p.calculateInverseKinematics(kuka_id, kuka_end_effector_idx, target_pos, target_orn)
    for j in range(num_joints):
        p.setJointMotorControl2(bodyIndex=kuka_id, jointIndex=j, controlMode=p.POSITION_CONTROL, targetPosition=joint_poses[j])

    # Control gripper opening/closing
    p.setJointMotorControl2(kuka_gripper_id, 4, p.POSITION_CONTROL, targetPosition=gripper_val*0.05, force=100)
    p.setJointMotorControl2(kuka_gripper_id, 6, p.POSITION_CONTROL, targetPosition=gripper_val*0.05, force=100)

    # Step simulation forward
    p.stepSimulation()

# Display the last frame as an image
plt.imshow(Image.fromarray(image))

# Close video recording and disconnect from PyBullet
vid.close()
p.disconnect()

````

## 2. "Open the Chests" Environment

image.png

The **Open the Chests** environment introduces **event-driven decision-making**, where the robot must **recognize patterns in an event stream** and take the appropriate action. The environment simulates a scenario where **various activities generate a stream of events**, each representing a **significant observation**.

The robot's goal is to:
- **Recognize event signatures** in the stream.
- **Associate events with specific activities**.
- **Use this information to open the correct chest** by taking the appropriate action (e.g., pressing a button).

This requires the robot to **learn how to interpret event sequences** and **respond accordingly**.


### **Challenge**
The key difficulty in this environment is handling **asynchronous and sequential events**. The robot must:
- **Interpret real-time event streams** where new information continuously arrives.
- **Recognize patterns** in events occurring at different points in time.
- **Distinguish relevant events** from noise in the event stream.
- **Identify and execute the correct action** based on learned patterns.

As event sequences become more **complex and interdependent**, the learning problem shifts from a simple **one-to-one mapping** to a **time-dependent decision-making** challenge.

### **Suggested Approach**
The **solution** to this problem can be approached at **multiple levels of complexity**:

#### **1. Simple Approach: Single Event Mapping**
- **Use a basic RL agent** that **associates the last seen event with a specific action**.
- The agent learns to **map a single event to an action** (e.g., pressing a button to open a chest).
- **Limitations:**  
  - Cannot handle **complex sequences** where multiple events must be considered.  
  - **Fails** if event history is required to make decisions.

#### **2. Intermediate Approach: Temporal Windows**
- Implement a **time window** where the agent **stores and analyzes recent events**.
- The agent can **recognize simple event sequences** by considering a **short history**.
- **Improvement over single-event mapping**, but still struggles with **long-range dependencies**.

#### **3. Advanced Approach: Transformers or Reccurence for Temporal Dependencies**
- **Utilize deep learning techniques** such as:
  - **Recurrent Neural Networks (RNNs)**
  - **Long Short-Term Memory (LSTMs)**
  - **Transformers (e.g., GPT or BERT-like models)**
- These models allow the agent to:
  - **Process complex event sequences** spanning many steps.
  - **Learn temporal relationships** and **predict upcoming events**.
  - **Generalize across different event patterns** instead of memorizing specific cases.




````python
# ------------------------------------------------------------------------------
# OpenTheChestsGym Environment Interaction & Event Visualization
# ------------------------------------------------------------------------------
# This script demonstrates:
# - How to initialize and interact with the OpenTheChestsGym environment.
# - How to extract event data from the environment.
# - How to visualize observed events in a timeline.
#
# The script follows these steps:
# 1. Define event types, attributes, and instructions.
# 2. Initialize the environment with the given configuration.
# 3. Run a loop to collect events while interacting with the environment.
# 4. Visualize the collected events using a timeline plot.
# ------------------------------------------------------------------------------

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import gym
from openthechests.src.OpenTheChestsGym import OpenTheChestsGym

# --------------------------------------------------------
# INITIALIZE THE ENVIRONMENT
# --------------------------------------------------------

# Define the list of event types and their attributes
all_event_types = ['A', 'B', 'C', 'D', 'E']
all_event_attributes = {
    'fg': ['red', 'blue', 'green', 'orange', 'pink'],
    'bg': ['red', 'blue', 'green', 'orange', 'pink']
}

# Define noise event types and attributes
all_noise_types = ['G', 'H', 'F']
all_noise_attributes = {
    'fg': ['yellow', 'purple', 'black'],
    'bg': ['yellow', 'purple', 'black']
}

# Define event instructions (simulating sequences of actions and dependencies)
instructions = [
    [
        {'command': 'delay', 'parameters': 10},
        {'command': 'instantiate', 'parameters': ('A', {'bg': 'blue', 'fg': 'blue'}, {'mu': 5, 'sigma': 2}), 'variable_name': 'distinct'},
        {'command': 'instantiate', 'parameters': ('E', {'bg': 'pink', 'fg': 'green'}, {'mu': 6, 'sigma': 2}), 'variable_name': 'e1'},
        {'command': 'instantiate', 'parameters': ('E', {'bg': 'red', 'fg': 'green'}, {'mu': 8, 'sigma': 0}), 'variable_name': 'e2'},
        {'command': 'met_by', 'parameters': ['e2', 'e1'], 'variable_name': 'e2', 'other': {}},
        {'command': 'after', 'parameters': ['distinct', 'e2'], 'variable_name': 'distinct', 'other': {'gap_dist': {'mu': 4, 'sigma': 1}}}
    ]
]

# Initialize the OpenTheChestsGym environment
env = OpenTheChestsGym(instructions=instructions,
                       all_event_types=all_event_types,
                       all_event_attributes=all_event_attributes,
                       all_noise_types=all_noise_types,
                       all_noise_attributes=all_noise_attributes,
                       verbose=False,
                       discrete=True, # If True actions are discrete, meaning that an action corresponds to the number of the box to open.
                                       # If False actions are binary vectors, meaning that a 0 at index i corresponds to pushing the buton i.
                       stb3=False)

# --------------------------------------------------------
# COLLECT EVENT OBSERVATIONS
# --------------------------------------------------------

def event_from_obs(obs):
    """
    Extracts event information from the environment observation.
    Converts indices into human-readable symbols and colors.
    """
    event = obs["context"]
    return {
        "symbol": all_event_types[event.type],  # Convert event index to actual event type
        "bg_color": all_event_attributes["bg"][event.attributes["bg"]],  # Convert bg index to color
        "symbol_color": all_event_attributes["fg"][event.attributes["fg"]],  # Convert fg index to color
        "start_time": event.start,
        "end_time": event.end
    }

# Store observed events
observed_events = []

# Reset environment and initialize observation
obs = env.reset()
observed_events.append(event_from_obs(obs))

# Run simulation loop to collect events
for step in range(10):
    action = env.action_space.sample()  # Fixed action (Can be randomized using env.action_space.sample())
    obs, reward, done, info = env.step(action)  # Take a step in the environment

    # Extract event from `obs["context"]`
    if "context" in obs and obs["context"] is not None:
        observed_events.append(event_from_obs(obs))

    print(f"Step {step+1}: Action={action}, Reward={reward}, Done={done}")

    if done:
        print("Environment reached a terminal state. Resetting...")
        obs = env.reset()

# --------------------------------------------------------
# VISUALIZE EVENT SEQUENCES
# --------------------------------------------------------

def draw_event_sequence_matplot(events, start_time=0, end_time=50, env_name = "Example"):
    """
    Visualizes the observed event sequence as a timeline, showing event occurrences over time.
    """
    # Create the figure and axis with a larger width
    fig, ax = plt.subplots(figsize=(15, 5))  # Increased width from 10 to 15

    last_event_end_times = []  # Track end times to avoid overlapping
    height = 1  # Height of each event bar

    for event in events:
        event_name = event["symbol"]
        start = event["start_time"]
        end = event["end_time"]
        color = event["bg_color"]
        text_color = event["symbol_color"]

        # Find the correct row for the event (avoid overlapping)
        line = 0
        while line < len(last_event_end_times):
            if start >= last_event_end_times[line]:  # If space is available in this line
                break
            line += 1
        if line == len(last_event_end_times):
            last_event_end_times.append(end)
        else:
            last_event_end_times[line] = end

        # Draw the event as a rectangle
        y_pos = line * (height + 0.5)
        rect = patches.Rectangle((start, y_pos), max(end - start, 0.1), height, color=color, alpha=0.7)  # Ensure non-zero width
        ax.add_patch(rect)

        # Label the event in the center of the rectangle
        ax.text(start + (end - start) / 2, y_pos + height / 2, event_name,
                horizontalalignment='center', verticalalignment='center',
                color=text_color, fontsize=12, fontweight='bold')

    # Set limits and labels
    ax.set_xlim(start_time, end_time)
    ax.set_ylim(0, len(last_event_end_times) * (height + 0.5))
    ax.set_xlabel("Time")
    ax.set_ylabel("Event Sequences")
    ax.set_title("Observed Event Timeline " + f'({env_name})')
    plt.show()


# Call the function to plot the observed event sequence
draw_event_sequence_matplot(observed_events, start_time=0, end_time=observed_events[-1]["end_time"])

# Close the environment
env.close()
````

### **Configurations**
The complexity of the problem can be adjusted based on how events are structured and how much information the robot must process. Below are the **four defined environments** with increasing levels of difficulty.

#### **1. Easy: 3 Chests, 1 Event Per Chest** (`OpenTheChests-v0`)
- Each **chest is associated with a single event**.
- The robot **only needs to react to a single event** to identify the correct chest.
- **No historical context or sequence tracking** is required.
- **Challenge Level: Minimal** â€“ A simple one-to-one mapping between an event and an action.
- Example:
  - **Event A** â†’ Open **Chest 1**  
  - **Event B** â†’ Open **Chest 2**  
  - **Event C** â†’ Open **Chest 3**  

#### **2. Medium: 3 Chests, 4 Events Per Chest** (`OpenTheChests-v1`)
- Each chest is associated with **a sequence of 4 events**.
- The robot **must recognize short event sequences** to determine the correct chest.
- **Some memory and sequence tracking** are required.
- **Challenge Level: Moderate** â€“ The robot must understand simple temporal dependencies.
- Example:
  - **Event A â†’ Event B â†’ Event C â†’ Event D** â†’ Open **Chest 1**  
  - **Event X â†’ Event Y â†’ Event Z â†’ Event W** â†’ Open **Chest 2**  
  - **Event M â†’ Event N â†’ Event O â†’ Event P** â†’ Open **Chest 3**  

#### **3. Hard: 3 Chests, 16 Events Per Chest** (`OpenTheChests-v2`)
- Each chest is associated with a **complex sequence of 16 events**.
- The robot **must track long-term dependencies** between events.
- **Deep memory and advanced pattern recognition** are necessary.
- **Challenge Level: High** â€“ The robot must handle **complex sequential dependencies** and **avoid distractions**.
- Example:
  - **Event red A â†’ after blue B â†’ during C â†’ after D â†’ ... â†’ Event P** â†’ Open **Chest 1**  
  - **Event X â†’ during red Y â†’ after pink Z â†’ during W â†’ ... â†’ Event O** â†’ Open **Chest 2**  
  - **Event M â†’ met_by N â†’ during O â†’ after P â†’ ... â†’ Event R** â†’ Open **Chest 3**  

### **Testing Different Complexity Levels**
These three configurations allow testing the **robotâ€™s ability to recognize patterns** in **low, medium, and high complexity cases**.

- **In the Easy environment**, the agent can use **basic reinforcement learning** to associate a **single event** with the correct action.
- **In the Medium environment**, the agent can use **simple memory storage** to correctly identify **event patterns**.
- **In the Hard environment**, the agent requires **advanced memory mechanisms (LSTMs, Transformers, or Attention-based models)** to detect **long event sequences**.

This setup allows for a **progressive increase in difficulty**, helping to evaluate **how different agents perform at various levels of complexity**.

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import gym
import random
from collections import deque
import matplotlib.pyplot as plt
from tqdm import tqdm

# --- Configuration ---
CONFIG = {
    "batch_size": 64,
    "learning_rate": 0.001,
    "epochs": 50,
    "hidden_dim": 128,
    "embed_dim": 64,
    "num_heads": 4,
    "num_encoder_layers": 4,
    "max_seq_length": 50,
    "device": torch.device("cuda" if torch.cuda.is_available() else "cpu"),
    "exploration_episodes": 500,
    "replay_buffer_size": 10000,
    "epsilon_start": 1.0,
    "epsilon_end": 0.1,
    "epsilon_decay": 0.995
}


# --- Fonctions d'aide ---
def event_from_obs_gym(obs, all_types, all_attributes):
    """Extrait les informations d'Ã©vÃ©nement de l'observation."""
    return {
        "symbol": all_types[obs["e_type"]],
        "bg_color": all_attributes["bg"][obs["bg"]],
        "symbol_color": all_attributes["fg"][obs["fg"]],
        "start_time": obs["start"][0],
        "end_time": obs["end"][0],
        "raw_type": obs["e_type"],
        "raw_bg": obs["bg"],
        "raw_fg": obs["fg"]
    }


# --- Classes pour le Dataset et le ModÃ¨le ---
class EventSequenceDataset(Dataset):
    """Dataset pour sÃ©quences d'Ã©vÃ©nements et actions correspondantes."""
    
    def __init__(self, sequences, actions, rewards, max_seq_length):
        self.sequences = sequences
        self.actions = actions
        self.rewards = rewards
        self.max_seq_length = max_seq_length
        
    def __len__(self):
        return len(self.actions)
    
    def __getitem__(self, idx):
        # RÃ©cupÃ©ration d'une sÃ©quence
        sequence = self.sequences[idx]
        action = self.actions[idx]
        reward = self.rewards[idx]
        
        # Padding pour avoir des sÃ©quences de mÃªme longueur
        if len(sequence) < self.max_seq_length:
            # Padding avec des Ã©vÃ©nements "vides"
            padding = [{"raw_type": -1, "raw_bg": -1, "raw_fg": -1, 
                        "relative_time": 0} for _ in range(self.max_seq_length - len(sequence))]
            padded_sequence = sequence + padding
        else:
            # Prendre seulement les derniers Ã©vÃ©nements
            padded_sequence = sequence[-self.max_seq_length:]
            
        # Convertir en tensors
        seq_tensor = torch.zeros((self.max_seq_length, 4), dtype=torch.float)
        mask_tensor = torch.zeros(self.max_seq_length, dtype=torch.bool)
        
        for i, event in enumerate(padded_sequence):
            if event["raw_type"] != -1:  # Ã‰vÃ©nement rÃ©el, pas du padding
                seq_tensor[i, 0] = event["raw_type"]
                seq_tensor[i, 1] = event["raw_bg"]
                seq_tensor[i, 2] = event["raw_fg"]
                seq_tensor[i, 3] = event["relative_time"]
                mask_tensor[i] = False  # Pas masquÃ©
            else:
                mask_tensor[i] = True  # MasquÃ© (padding)
        
        return {
            "sequence": seq_tensor,
            "mask": mask_tensor,
            "action": torch.tensor(action, dtype=torch.float),
            "reward": torch.tensor(reward, dtype=torch.float)
        }


class PositionalEncoding(nn.Module):
    """Encodage positionnel pour le Transformer."""
    
    def __init__(self, embed_dim, max_len=1000):
        super(PositionalEncoding, self).__init__()
        
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-np.log(10000.0) / embed_dim))
        
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]


class EventTransformer(nn.Module):
    """ModÃ¨le Transformer pour prÃ©dire les actions en fonction des sÃ©quences d'Ã©vÃ©nements."""
    
    def __init__(self, num_event_types, num_colors, embed_dim, hidden_dim, num_heads, 
                 num_encoder_layers, num_actions, max_seq_length):
        super(EventTransformer, self).__init__()
        
        # Embedding pour les types d'Ã©vÃ©nements et les couleurs
        self.event_embedding = nn.Embedding(num_event_types + 1, embed_dim // 2)  # +1 pour le padding
        self.bg_embedding = nn.Embedding(num_colors + 1, embed_dim // 4)
        self.fg_embedding = nn.Embedding(num_colors + 1, embed_dim // 4)
        
        # Couche linÃ©aire pour traiter le temps relatif
        self.time_proj = nn.Linear(1, embed_dim)
        
        # Encodage positionnel
        self.pos_encoder = PositionalEncoding(embed_dim, max_seq_length)
        
        # Encoder du Transformer
        encoder_layers = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads, 
                                                   dim_feedforward=hidden_dim, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_encoder_layers)
        
        # Couches de sortie
        self.fc1 = nn.Linear(embed_dim, hidden_dim)
        self.dropout = nn.Dropout(0.1)
        self.fc2 = nn.Linear(hidden_dim, num_actions)
        
    def forward(self, x, mask=None):
        # x a la forme [batch_size, seq_len, 4] (type, bg, fg, time)
        
        # Extraction des composants
        event_type = x[:, :, 0].long()
        bg_color = x[:, :, 1].long()
        fg_color = x[:, :, 2].long()
        rel_time = x[:, :, 3].unsqueeze(-1)
        
        # Application des embeddings
        type_emb = self.event_embedding(event_type)
        bg_emb = self.bg_embedding(bg_color)
        fg_emb = self.fg_embedding(fg_color)
        time_emb = self.time_proj(rel_time)
        
        # ConcatÃ©nation des embeddings
        combined = torch.cat([type_emb, bg_emb, fg_emb], dim=2)
        combined = combined + time_emb
        
        # Encodage positionnel
        combined = self.pos_encoder(combined)
        
        # Application du masque si fourni
        if mask is not None:
            # Transformer les masques pour qu'ils correspondent aux attentes de PyTorch
            src_key_padding_mask = mask
            
            # Appliquer l'encodeur Transformer avec le masque
            output = self.transformer_encoder(combined, src_key_padding_mask=src_key_padding_mask)
        else:
            output = self.transformer_encoder(combined)
        
        # Prendre la reprÃ©sentation moyenne de la sÃ©quence pour la dÃ©cision
        # Alternativement, on pourrait utiliser juste le dernier Ã©vÃ©nement non masquÃ©
        seq_representation = output.mean(dim=1)
        
        # Couches fully connected
        x = self.fc1(seq_representation)
        x = torch.relu(x)
        x = self.dropout(x)
        logits = self.fc2(x)
        
        return logits


# --- Collecte de donnÃ©es ---
def collect_data(env_name, all_types, all_attributes, num_episodes=500, epsilon_start=1.0, epsilon_end=0.1, epsilon_decay=0.995):
    """Collecte des donnÃ©es en explorant l'environnement avec une stratÃ©gie epsilon-greedy."""
    env = gym.make(env_name)
    num_actions = env.action_space.n
    
    # Initialisation du buffer de replay
    replay_buffer = deque(maxlen=CONFIG["replay_buffer_size"])
    
    # Exploration avec epsilon-greedy
    epsilon = epsilon_start
    
    for episode in tqdm(range(num_episodes), desc="Collecte des donnÃ©es"):
        obs = env.reset()
        done = False
        episode_sequence = []
        
        # Temps initial pour calculer les temps relatifs
        initial_time = obs["start"][0]
        
        while not done:
            # Ajouter l'Ã©vÃ©nement Ã  la sÃ©quence
            event = event_from_obs_gym(obs, all_types, all_attributes)
            event["relative_time"] = event["start_time"] - initial_time
            episode_sequence.append(event)
            
            # SÃ©lection d'action epsilon-greedy
            if random.random() < epsilon:
                # CrÃ©er un vecteur d'action avec des 0 et un 1 Ã  un index alÃ©atoire
                action = [0] * num_actions
                if env_name.startswith("OpenTheChests"):
                    # Pour OpenTheChests, action est un vecteur binaire
                    selected_box = random.randint(0, num_actions - 1)
                    action[selected_box] = 1
            else:
                # Action par dÃ©faut (tous Ã  0)
                action = [0] * num_actions
                # SÃ©lectionner une boÃ®te alÃ©atoirement pour l'exploitation
                selected_box = random.randint(0, num_actions - 1)
                action[selected_box] = 1
            
            # ExÃ©cuter l'action
            next_obs, reward, done, info = env.step(action)
            
            # Stocker l'expÃ©rience si la sÃ©quence a une longueur suffisante
            if len(episode_sequence) > 1:
                replay_buffer.append((
                    list(episode_sequence),  # Copie de la sÃ©quence actuelle
                    action,  # Action sous forme de vecteur
                    reward
                ))
            
            obs = next_obs
        
        # DÃ©crÃ©menter epsilon
        epsilon = max(epsilon_end, epsilon * epsilon_decay)
    
    env.close()
    return replay_buffer


# --- PrÃ©paration des donnÃ©es et entraÃ®nement ---
def prepare_dataset(replay_buffer, max_seq_length):
    """PrÃ©pare le dataset Ã  partir du buffer de replay."""
    sequences = []
    actions = []
    rewards = []
    
    for seq, action, reward in replay_buffer:
        sequences.append(seq)
        actions.append(action)
        rewards.append(reward)
    
    return EventSequenceDataset(sequences, actions, rewards, max_seq_length)


def train_model(model, train_loader, optimizer, criterion, device, epochs=10):
    """EntraÃ®ne le modÃ¨le sur le dataset."""
    model.train()
    
    epoch_losses = []
    
    for epoch in range(epochs):
        total_loss = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            sequences = batch["sequence"].to(device)
            masks = batch["mask"].to(device)
            actions = batch["action"].to(device)
            rewards = batch["reward"].to(device)
            
            # Forward pass
            logits = model(sequences, masks)
            
            # MSE pour actions vectorielles
            loss = criterion(logits, actions)
            
            # Si la rÃ©compense est nÃ©gative, on peut rÃ©duire l'importance de cette erreur
            loss = loss * torch.where(rewards > 0, 1.0, 0.5).unsqueeze(1)
            loss = loss.mean()
            
            # Backward et optimisation
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        
        epoch_loss = total_loss / len(train_loader)
        epoch_losses.append(epoch_loss)
        
        print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}")
    
    # Tracer la courbe de perte
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, epochs + 1), epoch_losses)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.grid(True)
    plt.show()
    
    return model


def evaluate_model(model, env_name, all_types, all_attributes, max_seq_length, device, num_episodes=100):
    """Ã‰value le modÃ¨le entraÃ®nÃ© sur l'environnement."""
    env = gym.make(env_name)
    num_actions = env.action_space.n
    model.eval()
    
    total_rewards = []
    
    for episode in tqdm(range(num_episodes), desc="Ã‰valuation"):
        obs = env.reset()
        done = False
        episode_reward = 0
        episode_sequence = []
        
        # Temps initial pour calculer les temps relatifs
        initial_time = obs["start"][0]
        
        while not done:
            # Ajouter l'Ã©vÃ©nement Ã  la sÃ©quence
            event = event_from_obs_gym(obs, all_types, all_attributes)
            event["relative_time"] = event["start_time"] - initial_time
            episode_sequence.append(event)
            
            # PrÃ©parer l'entrÃ©e pour le modÃ¨le
            if len(episode_sequence) < max_seq_length:
                padding = [{"raw_type": -1, "raw_bg": -1, "raw_fg": -1, 
                           "relative_time": 0} for _ in range(max_seq_length - len(episode_sequence))]
                padded_sequence = episode_sequence + padding
            else:
                padded_sequence = episode_sequence[-max_seq_length:]
            
            # Convertir en tensors
            seq_tensor = torch.zeros((1, max_seq_length, 4), dtype=torch.float)
            mask_tensor = torch.zeros((1, max_seq_length), dtype=torch.bool)
            
            for i, event in enumerate(padded_sequence):
                if event["raw_type"] != -1:  # Ã‰vÃ©nement rÃ©el, pas du padding
                    seq_tensor[0, i, 0] = event["raw_type"]
                    seq_tensor[0, i, 1] = event["raw_bg"]
                    seq_tensor[0, i, 2] = event["raw_fg"]
                    seq_tensor[0, i, 3] = event["relative_time"]
                    mask_tensor[0, i] = False  # Pas masquÃ©
                else:
                    mask_tensor[0, i] = True  # MasquÃ© (padding)
            
            # PrÃ©dire l'action
            with torch.no_grad():
                seq_tensor = seq_tensor.to(device)
                mask_tensor = mask_tensor.to(device)
                logits = model(seq_tensor, mask_tensor)
                
                # Convertir les logits en vecteur d'action
                action_vector = [0] * num_actions
                selected_box = torch.argmax(logits[0]).item()
                action_vector[selected_box] = 1
            
            # ExÃ©cuter l'action
            next_obs, reward, done, info = env.step(action_vector)
            episode_reward += reward
            
            obs = next_obs
        
        total_rewards.append(episode_reward)
    
    env.close()
    
    # Afficher les rÃ©sultats
    avg_reward = sum(total_rewards) / len(total_rewards)
    print(f"RÃ©compense moyenne sur {num_episodes} Ã©pisodes: {avg_reward:.2f}")
    
    return avg_reward, total_rewards


# --- Programme principal ---
def main():
    from register_envs import all_types, all_attributes, register_custom_envs
    register_custom_envs()
    
    env_name = "OpenTheChests-v2"  # Utiliser l'environnement difficile
    print(f"Utilisation de l'environnement: {env_name}")
    
    # Configuration
    device = CONFIG["device"]
    print(f"Utilisation du pÃ©riphÃ©rique: {device}")
    
    # Collecte des donnÃ©es
    print("Collecte des donnÃ©es d'entraÃ®nement...")
    replay_buffer = collect_data(
        env_name, 
        all_types, 
        all_attributes,
        num_episodes=CONFIG["exploration_episodes"],
        epsilon_start=CONFIG["epsilon_start"],
        epsilon_end=CONFIG["epsilon_end"],
        epsilon_decay=CONFIG["epsilon_decay"]
    )
    
    print(f"DonnÃ©es collectÃ©es: {len(replay_buffer)} Ã©chantillons")
    
    # PrÃ©paration du dataset
    dataset = prepare_dataset(replay_buffer, CONFIG["max_seq_length"])
    
    # Division en ensembles d'entraÃ®nement et de validation
    train_size = int(0.8 * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_dataset, batch_size=CONFIG["batch_size"], shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=CONFIG["batch_size"])
    
    print(f"Taille de l'ensemble d'entraÃ®nement: {len(train_dataset)}")
    print(f"Taille de l'ensemble de validation: {len(val_dataset)}")
    
    # Initialisation du modÃ¨le
    env = gym.make(env_name)
    num_actions = env.action_space.n
    env.close()
    
    model = EventTransformer(
        num_event_types=len(all_types),
        num_colors=len(all_attributes["bg"]),
        embed_dim=CONFIG["embed_dim"],
        hidden_dim=CONFIG["hidden_dim"],
        num_heads=CONFIG["num_heads"],
        num_encoder_layers=CONFIG["num_encoder_layers"],
        num_actions=num_actions,
        max_seq_length=CONFIG["max_seq_length"]
    ).to(device)
    
    # Affichage du modÃ¨le
    print(model)
    
    # Optimiseur et fonction de perte
    optimizer = optim.Adam(model.parameters(), lr=CONFIG["learning_rate"])
    criterion = nn.MSELoss(reduction='none')  # MSE pour actions vectorielles
    
    # EntraÃ®nement
    print("DÃ©but de l'entraÃ®nement...")
    trained_model = train_model(
        model,
        train_loader,
        optimizer,
        criterion,
        device,
        epochs=CONFIG["epochs"]
    )
    
    # Ã‰valuation
    print("Ã‰valuation du modÃ¨le...")
    avg_reward, rewards = evaluate_model(
        trained_model,
        env_name,
        all_types,
        all_attributes,
        CONFIG["max_seq_length"],
        device
    )
    
    # Sauvegarde du modÃ¨le
    torch.save(trained_model.state_dict(), "event_transformer_model.pt")
    print("ModÃ¨le sauvegardÃ© sous 'event_transformer_model.pt'")
    
    # Visualisation des rÃ©compenses
    plt.figure(figsize=(10, 5))
    plt.plot(rewards)
    plt.xlabel("Ã‰pisode")
    plt.ylabel("RÃ©compense")
    plt.title("RÃ©compenses par Ã©pisode")
    plt.grid(True)
    plt.show()


if __name__ == "__main__":
    main()
````

````python
def collect_data(env_name, all_types, all_attributes, num_episodes=500, epsilon_start=1.0, epsilon_end=0.05, epsilon_decay=0.99):
    """Improved data collection with better exploration strategy."""
    env = gym.make(env_name)
    num_actions = env.action_space.n
    
    # Buffer with prioritized experience replay
    replay_buffer = deque(maxlen=CONFIG["replay_buffer_size"])
    rewards_history = []  # Track rewards for adaptive exploration
    
    # Use a decaying epsilon but with adaptive adjustment
    epsilon = epsilon_start
    
    for episode in tqdm(range(num_episodes), desc="Collecting data"):
        obs = env.reset()
        done = False
        episode_sequence = []
        episode_reward = 0
        
        # Time tracking
        initial_time = obs["start"][0]
        
        while not done:
            # Add event to sequence with better feature extraction
            event = event_from_obs_gym(obs, all_types, all_attributes)
            event["relative_time"] = event["start_time"] - initial_time
            # Add more features like time since last similar event
            if len(episode_sequence) > 0:
                for prev_event in reversed(episode_sequence):
                    if prev_event["raw_type"] == event["raw_type"]:
                        event["time_since_similar"] = event["relative_time"] - prev_event["relative_time"]
                        break
                else:
                    event["time_since_similar"] = event["relative_time"]
            else:
                event["time_since_similar"] = event["relative_time"]
            
            episode_sequence.append(event)
            
            # Epsilon-greedy with smarter exploration
            if random.random() < epsilon:
                # Use previous rewards to bias action selection when exploring
                action = [0] * num_actions
                if len(rewards_history) > 10:  # Have some history
                    # Bias toward actions that worked before
                    probs = np.ones(num_actions) * 0.1  # Base probability
                    for i in range(num_actions):
                        # Increase probability for actions that led to rewards
                        temp_action = [0] * num_actions
                        temp_action[i] = 1
                        action_str = str(temp_action)
                        if action_str in rewards_history:
                            probs[i] += 0.3 * rewards_history[action_str]
                    probs = probs / probs.sum()  # Normalize
                    selected_box = np.random.choice(num_actions, p=probs)
                else:
                    selected_box = random.randint(0, num_actions - 1)
                action[selected_box] = 1
            else:
                # Make best guess based on patterns seen so far
                action = [0] * num_actions
                selected_box = random.randint(0, num_actions - 1)  # Will be replaced with model prediction later
                action[selected_box] = 1
            
            # Execute action
            next_obs, reward, done, info = env.step(action)
            episode_reward += reward
            
            # Store experience with more context
            if len(episode_sequence) > 1:
                # Create a copy with all relevant information
                replay_buffer.append((
                    list(episode_sequence),  # Current sequence
                    action,
                    reward,
                    episode_reward  # Track cumulative reward
                ))
                
                # Track successful actions
                if reward > 0:
                    if not hasattr(collect_data, "rewards_history"):
                        collect_data.rewards_history = {}
                    action_str = str(action)
                    if action_str not in collect_data.rewards_history:
                        collect_data.rewards_history[action_str] = 0
                    collect_data.rewards_history[action_str] += reward
            
            obs = next_obs
        
        # Adaptive epsilon decay - slow down decay if rewards are low
        avg_reward = episode_reward / len(episode_sequence) if episode_sequence else -1
        if avg_reward < -0.5:  # If doing poorly, reduce decay rate
            epsilon = max(epsilon_end, epsilon * (epsilon_decay + 0.005))
        else:
            epsilon = max(epsilon_end, epsilon * epsilon_decay)
            
        # Track episode performance
        rewards_history.append(episode_reward)
    
    env.close()
    return replay_buffer
````

````python
class EnhancedEventTransformer(nn.Module):
    """Enhanced Transformer model with better feature processing and attention mechanisms."""
    
    def __init__(self, num_event_types, num_colors, embed_dim, hidden_dim, num_heads, 
                 num_encoder_layers, num_actions, max_seq_length):
        super(EnhancedEventTransformer, self).__init__()
        
        # Feature-specific embeddings with larger dimensions
        self.event_embedding = nn.Embedding(num_event_types + 1, embed_dim // 2)
        self.bg_embedding = nn.Embedding(num_colors + 1, embed_dim // 4)
        self.fg_embedding = nn.Embedding(num_colors + 1, embed_dim // 4)
        
        # Better time processing with non-linear transformation
        self.time_encoder = nn.Sequential(
            nn.Linear(2, embed_dim // 2),  # Process both absolute and relative time
            nn.ReLU(),
            nn.Linear(embed_dim // 2, embed_dim),
        )
        
        # Positional encoding
        self.pos_encoder = PositionalEncoding(embed_dim, max_seq_length)
        
        # Improved transformer with higher capacity
        encoder_layers = nn.TransformerEncoderLayer(
            d_model=embed_dim, 
            nhead=num_heads,
            dim_feedforward=hidden_dim * 2,  # Increased capacity
            dropout=0.2,  # Stronger regularization
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_encoder_layers)
        
        # Advanced output processing
        self.output_processor = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, num_actions)
        )
        
        # Additional layer to estimate value for each action
        self.value_head = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, 1)
        )
        
    def forward(self, x, mask=None):
        # x has shape [batch_size, seq_len, features]
        
        # Extract components
        event_type = x[:, :, 0].long()
        bg_color = x[:, :, 1].long()
        fg_color = x[:, :, 2].long()
        rel_time = x[:, :, 3].unsqueeze(-1)
        time_since_similar = x[:, :, 4].unsqueeze(-1) if x.shape[2] > 4 else rel_time.clone()
        
        # Time processing
        time_features = torch.cat([rel_time, time_since_similar], dim=2)
        time_emb = self.time_encoder(time_features)
        
        # Apply embeddings
        type_emb = self.event_embedding(event_type)
        bg_emb = self.bg_embedding(bg_color)
        fg_emb = self.fg_embedding(fg_color)
        
        # Combine embeddings
        combined = torch.cat([type_emb, bg_emb, fg_emb], dim=2)
        combined = combined + time_emb
        
        # Positional encoding
        combined = self.pos_encoder(combined)
        
        # Apply transformer with mask if provided
        if mask is not None:
            output = self.transformer_encoder(combined, src_key_padding_mask=mask)
        else:
            output = self.transformer_encoder(combined)
        
        # Extract sequence representation - use attention over full sequence
        # instead of simple mean pooling
        attn_weights = torch.softmax(self.value_head(output), dim=1)
        seq_representation = torch.sum(output * attn_weights, dim=1)
        
        # Generate action outputs
        action_logits = self.output_processor(seq_representation)
        
        return action_logits
````

````python
def train_model(model, train_loader, val_loader, optimizer, device, epochs=10):
    """Enhanced training process with curriculum learning and better monitoring."""
    model.train()
    
    # Track metrics
    epoch_losses = []
    val_losses = []
    best_val_loss = float('inf')
    best_model = None
    patience = 5
    patience_counter = 0
    
    # Define loss function with reward weighting
    def compute_loss(logits, actions, rewards):
        # MSE loss for actions
        mse_loss = nn.MSELoss(reduction='none')(logits, actions)
        
        # Weight loss by reward - higher rewards get more importance
        reward_weights = torch.exp(rewards / 10.0).unsqueeze(1)  # Scale to avoid exploding values
        weighted_loss = mse_loss * reward_weights
        return weighted_loss.mean()
    
    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=3, verbose=True
    )
    
    for epoch in range(epochs):
        total_loss = 0
        
        # Training loop
        model.train()
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            sequences = batch["sequence"].to(device)
            masks = batch["mask"].to(device)
            actions = batch["action"].to(device)
            rewards = batch["reward"].to(device)
            
            # Forward pass
            logits = model(sequences, masks)
            
            # Calculate loss
            loss = compute_loss(logits, actions, rewards)
            
            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            
            # Gradient clipping to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()
            total_loss += loss.item()
        
        epoch_loss = total_loss / len(train_loader)
        epoch_losses.append(epoch_loss)
        
        # Validation loop
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for batch in val_loader:
                sequences = batch["sequence"].to(device)
                masks = batch["mask"].to(device)
                actions = batch["action"].to(device)
                rewards = batch["reward"].to(device)
                
                logits = model(sequences, masks)
                loss = compute_loss(logits, actions, rewards)
                val_loss += loss.item()
        
        val_loss = val_loss / len(val_loader)
        val_losses.append(val_loss)
        
        # Update learning rate based on validation loss
        scheduler.step(val_loss)
        
        # Save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model = copy.deepcopy(model.state_dict())
            patience_counter = 0
        else:
            patience_counter += 1
        
        # Early stopping
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break
        
        print(f"Epoch {epoch+1}, Train Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}")
    
    # Load best model
    if best_model is not None:
        model.load_state_dict(best_model)
    
    # Plot losses
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(epoch_losses) + 1), epoch_losses, label='Train Loss')
    plt.plot(range(1, len(val_losses) + 1), val_losses, label='Val Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training & Validation Loss")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return model
````

````python
class ImprovedEventSequenceDataset(Dataset):
    """Enhanced dataset with better feature extraction and preprocessing."""
    
    def __init__(self, sequences, actions, rewards, max_seq_length):
        self.sequences = sequences
        self.actions = actions
        self.rewards = rewards
        self.max_seq_length = max_seq_length
        
        # Preprocess to normalize features
        self.preprocess_sequences()
        
    def preprocess_sequences(self):
        """Preprocess sequences to normalize and extract additional features."""
        # Find min/max for normalization
        max_time = 0
        for seq in self.sequences:
            for event in seq:
                max_time = max(max_time, event["relative_time"])
                if "time_since_similar" in event:
                    max_time = max(max_time, event["time_since_similar"])
        
        # Apply normalization
        for seq in self.sequences:
            for event in seq:
                # Normalize time values
                event["relative_time"] = event["relative_time"] / max(1, max_time)
                if "time_since_similar" in event:
                    event["time_since_similar"] = event["time_since_similar"] / max(1, max_time)
                else:
                    event["time_since_similar"] = event["relative_time"]
    
    def __len__(self):
        return len(self.actions)
    
    def __getitem__(self, idx):
        # Get sequence
        sequence = self.sequences[idx]
        action = self.actions[idx]
        reward = self.rewards[idx]
        
        # Handle padding
        if len(sequence) < self.max_seq_length:
            # Pad with empty events
            padding = [{"raw_type": -1, "raw_bg": -1, "raw_fg": -1, 
                        "relative_time": 0, "time_since_similar": 0} 
                       for _ in range(self.max_seq_length - len(sequence))]
            padded_sequence = sequence + padding
        else:
            # Use most recent events
            padded_sequence = sequence[-self.max_seq_length:]
            
        # Convert to tensors with additional features
        seq_tensor = torch.zeros((self.max_seq_length, 5), dtype=torch.float)  # 5 features now
        mask_tensor = torch.zeros(self.max_seq_length, dtype=torch.bool)
        
        for i, event in enumerate(padded_sequence):
            if event["raw_type"] != -1:  # Real event, not padding
                seq_tensor[i, 0] = event["raw_type"]
                seq_tensor[i, 1] = event["raw_bg"]
                seq_tensor[i, 2] = event["raw_fg"]
                seq_tensor[i, 3] = event["relative_time"]
                seq_tensor[i, 4] = event.get("time_since_similar", event["relative_time"])
                mask_tensor[i] = False
            else:
                mask_tensor[i] = True
        
        return {
            "sequence": seq_tensor,
            "mask": mask_tensor,
            "action": torch.tensor(action, dtype=torch.float),
            "reward": torch.tensor(reward, dtype=torch.float)
        }
````

````python
def evaluate_model(model, env_name, all_types, all_attributes, max_seq_length, device, num_episodes=100):
    """Enhanced evaluation with better metrics and visualizations."""
    env = gym.make(env_name)
    num_actions = env.action_space.n
    model.eval()
    
    total_rewards = []
    episode_lengths = []
    action_distributions = np.zeros(num_actions)
    
    for episode in tqdm(range(num_episodes), desc="Evaluation"):
        obs = env.reset()
        done = False
        episode_reward = 0
        episode_sequence = []
        steps = 0
        
        # Initial time
        initial_time = obs["start"][0]
        max_time_seen = 0  # For normalization
        
        while not done:
            # Add event to sequence with enhanced features
            event = event_from_obs_gym(obs, all_types, all_attributes)
            event["relative_time"] = event["start_time"] - initial_time
            
            # Track time since similar event
            if len(episode_sequence) > 0:
                for prev_event in reversed(episode_sequence):
                    if prev_event["raw_type"] == event["raw_type"]:
                        event["time_since_similar"] = event["relative_time"] - prev_event["relative_time"]
                        break
                else:
                    event["time_since_similar"] = event["relative_time"]
            else:
                event["time_since_similar"] = event["relative_time"]
            
            # Normalize time features based on max time seen so far
            max_time_seen = max(max_time_seen, event["relative_time"], event["time_since_similar"])
            event["relative_time"] = event["relative_time"] / max(1, max_time_seen)
            event["time_since_similar"] = event["time_since_similar"] / max(1, max_time_seen)
            
            episode_sequence.append(event)
            
            # Prepare input for model
            if len(episode_sequence) < max_seq_length:
                padding = [{"raw_type": -1, "raw_bg": -1, "raw_fg": -1, 
                           "relative_time": 0, "time_since_similar": 0} 
                          for _ in range(max_seq_length - len(episode_sequence))]
                padded_sequence = episode_sequence + padding
            else:
                padded_sequence = episode_sequence[-max_seq_length:]
            
            # Convert to tensors
            seq_tensor = torch.zeros((1, max_seq_length, 5), dtype=torch.float)  # Now 5 features
            mask_tensor = torch.zeros((1, max_seq_length), dtype=torch.bool)
            
            for i, event in enumerate(padded_sequence):
                if event["raw_type"] != -1:  # Real event
                    seq_tensor[0, i, 0] = event["raw_type"]
                    seq_tensor[0, i, 1] = event["raw_bg"]
                    seq_tensor[0, i, 2] = event["raw_fg"]
                    seq_tensor[0, i, 3] = event["relative_time"]
                    seq_tensor[0, i, 4] = event.get("time_since_similar", event["relative_time"])
                    mask_tensor[0, i] = False
                else:
                    mask_tensor[0, i] = True
            
            # Predict action with confidence
            with torch.no_grad():
                seq_tensor = seq_tensor.to(device)
                mask_tensor = mask_tensor.to(device)
                logits = model(seq_tensor, mask_tensor)
                
                # Generate action with some exploration during evaluation
                # This can help find better solutions
                if random.random() < 0.05:  # 5% exploration
                    selected_box = random.randint(0, num_actions - 1)
                else:
                    selected_box = torch.argmax(logits[0]).item()
                
                action_vector = [0] * num_actions
                action_vector[selected_box] = 1
                action_distributions[selected_box] += 1
            
            # Execute action
            next_obs, reward, done, info = env.step(action_vector)
            episode_reward += reward
            steps += 1
            
            obs = next_obs
        
        total_rewards.append(episode_reward)
        episode_lengths.append(steps)
    
    env.close()
    
    # Calculate statistics
    avg_reward = sum(total_rewards) / len(total_rewards)
    avg_length = sum(episode_lengths) / len(episode_lengths)
    action_probs = action_distributions / action_distributions.sum()
    
    # Print results
    print(f"Average reward over {num_episodes} episodes: {avg_reward:.2f}")
    print(f"Average episode length: {avg_length:.2f} steps")
    print(f"Action distribution: {action_probs}")
    
    # Visualize results
    plt.figure(figsize=(15, 5))
    
    # Plot rewards
    plt.subplot(1, 3, 1)
    plt.plot(total_rewards)
    plt.axhline(avg_reward, color='r', linestyle='--', label=f'Avg: {avg_reward:.2f}')
    plt.xlabel("Episode")
    plt.ylabel("Reward")
    plt.title("Episode Rewards")
    plt.legend()
    
    # Plot episode lengths
    plt.subplot(1, 3, 2)
    plt.plot(episode_lengths)
    plt.axhline(avg_length, color='r', linestyle='--', label=f'Avg: {avg_length:.2f}')
    plt.xlabel("Episode")
    plt.ylabel("Steps")
    plt.title("Episode Lengths")
    plt.legend()
    
    # Plot action distribution
    plt.subplot(1, 3, 3)
    plt.bar(range(num_actions), action_probs)
    plt.xlabel("Action")
    plt.ylabel("Frequency")
    plt.title("Action Distribution")
    
    plt.tight_layout()
    plt.show()
    
    return avg_reward, total_rewards, action_probs
````

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import gym
import random
from collections import deque, namedtuple
import matplotlib.pyplot as plt
from tqdm import tqdm
import copy

# Define a proper configuration
CONFIG = {
    "batch_size": 64,
    "learning_rate": 0.0005,
    "gamma": 0.99,  # Discount factor
    "target_update": 10,  # How often to update target network
    "memory_size": 10000,  # Replay buffer size
    "hidden_dim": 256,
    "embed_dim": 128,
    "num_heads": 8,
    "num_encoder_layers": 6,
    "max_seq_length": 50,
    "device": torch.device("cuda" if torch.cuda.is_available() else "cpu"),
    "epsilon_start": 1.0,
    "epsilon_end": 0.05,
    "epsilon_decay": 0.995
}

# Modified event processor function to handle different observation formats
def event_from_obs_gym(obs, all_types, all_attributes):
    """Extract event information from gym observation."""
    try:
        # Check if observation is a dictionary
        if isinstance(obs, dict):
            return {
                "symbol": all_types[obs["e_type"]] if "e_type" in obs else "unknown",
                "bg_color": all_attributes["bg"][obs["bg"]] if "bg" in obs else "unknown",
                "symbol_color": all_attributes["fg"][obs["fg"]] if "fg" in obs else "unknown",
                "start_time": obs["start"][0] if "start" in obs and isinstance(obs["start"], (list, np.ndarray)) else 0,
                "end_time": obs["end"][0] if "end" in obs and isinstance(obs["end"], (list, np.ndarray)) else 0,
                "raw_type": obs["e_type"] if "e_type" in obs else 0,
                "raw_bg": obs["bg"] if "bg" in obs else 0,
                "raw_fg": obs["fg"] if "fg" in obs else 0
            }
        # Handle list or array type observations
        elif isinstance(obs, (list, np.ndarray)):
            # Assuming structure: [e_type, bg, fg, start_time, end_time]
            # Adjust indices based on actual observation structure
            return {
                "raw_type": int(obs[0]) if len(obs) > 0 else 0,
                "raw_bg": int(obs[1]) if len(obs) > 1 else 0,
                "raw_fg": int(obs[2]) if len(obs) > 2 else 0,
                "start_time": float(obs[3]) if len(obs) > 3 else 0,
                "end_time": float(obs[4]) if len(obs) > 4 else 0,
                "symbol": all_types[int(obs[0])] if len(obs) > 0 and int(obs[0]) < len(all_types) else "unknown",
                "bg_color": all_attributes["bg"][int(obs[1])] if len(obs) > 1 and int(obs[1]) < len(all_attributes["bg"]) else "unknown",
                "symbol_color": all_attributes["fg"][int(obs[2])] if len(obs) > 2 and int(obs[2]) < len(all_attributes["fg"]) else "unknown"
            }
        else:
            # Fallback for other observation types
            print(f"Warning: Unexpected observation type: {type(obs)}")
            return {
                "raw_type": 0,
                "raw_bg": 0,
                "raw_fg": 0,
                "start_time": 0,
                "end_time": 0,
                "symbol": "unknown",
                "bg_color": "unknown",
                "symbol_color": "unknown"
            }
    except Exception as e:
        print(f"Error processing observation: {e}")
        print(f"Observation: {obs}")
        # Return a default event
        return {
            "raw_type": 0,
            "raw_bg": 0,
            "raw_fg": 0,
            "start_time": 0,
            "end_time": 0,
            "symbol": "unknown",
            "bg_color": "unknown",
            "symbol_color": "unknown"
        }

# Experience replay memory
Experience = namedtuple('Experience', ('state', 'action', 'next_state', 'reward', 'done'))

class ReplayMemory:
    def __init__(self, capacity):
        self.memory = deque(maxlen=capacity)
    
    def push(self, *args):
        self.memory.append(Experience(*args))
    
    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_seq_length=100):
        super().__init__()
        pe = torch.zeros(max_seq_length, d_model)
        position = torch.arange(0, max_seq_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * -(np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))
        
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]

class EventEncoder(nn.Module):
    """Encodes individual events into embeddings."""
    def __init__(self, num_event_types, num_colors, embed_dim):
        super().__init__()
        # Embeddings for discrete features
        self.type_embedding = nn.Embedding(num_event_types + 1, embed_dim // 3)  # +1 for padding
        self.bg_embedding = nn.Embedding(num_colors + 1, embed_dim // 3)
        self.fg_embedding = nn.Embedding(num_colors + 1, embed_dim // 3)
        
        # Processing for time features
        self.time_encoder = nn.Sequential(
            nn.Linear(2, embed_dim),
            nn.ReLU(),
            nn.Linear(embed_dim, embed_dim)
        )
        
        # Final projection to ensure consistent dimension
        self.proj = nn.Linear(embed_dim + (embed_dim // 3) * 3, embed_dim)
        self.norm = nn.LayerNorm(embed_dim)
        
    def forward(self, event_features):
        # event_features: [batch_size, seq_len, features]
        # features are: [type, bg, fg, start_time, end_time]
        
        event_type = event_features[:, :, 0].long()
        bg_color = event_features[:, :, 1].long()
        fg_color = event_features[:, :, 2].long()
        
        # Get embeddings for discrete features
        type_emb = self.type_embedding(event_type)
        bg_emb = self.bg_embedding(bg_color)
        fg_emb = self.fg_embedding(fg_color)
        
        # Process time features
        time_features = event_features[:, :, 3:5]  # start_time and end_time
        time_emb = self.time_encoder(time_features)
        
        # Combine all embeddings
        combined = torch.cat([type_emb, bg_emb, fg_emb, time_emb], dim=2)
        output = self.proj(combined)
        return self.norm(output)

class TransformerDQN(nn.Module):
    def __init__(self, num_event_types, num_colors, embed_dim, hidden_dim, 
                num_heads, num_encoder_layers, num_actions, max_seq_length):
        super().__init__()
        
        # Event encoder
        self.event_encoder = EventEncoder(num_event_types, num_colors, embed_dim)
        
        # Positional encoding
        self.pos_encoder = PositionalEncoding(embed_dim, max_seq_length)
        
        # Transformer encoder
        encoder_layers = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=num_heads,
            dim_feedforward=hidden_dim,
            dropout=0.1,
            batch_first=True
        )
        self.transformer = nn.TransformerEncoder(
            encoder_layers,
            num_layers=num_encoder_layers
        )
        
        # Q-value prediction
        self.q_net = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, num_actions)
        )
        
        # Advantage and value streams for dueling DQN architecture
        self.advantage_stream = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, num_actions)
        )
        
        self.value_stream = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, 1)
        )
        
    def forward(self, x, mask=None):
        # x shape: [batch_size, seq_len, features]
        
        # Encode individual events
        encoded_events = self.event_encoder(x)
        
        # Add positional encoding
        encoded_events = self.pos_encoder(encoded_events)
        
        # Apply transformer with masking if provided
        if mask is not None:
            transformer_out = self.transformer(encoded_events, src_key_padding_mask=mask)
        else:
            transformer_out = self.transformer(encoded_events)
        
        # Global representation (using mean)
        global_rep = transformer_out.mean(dim=1)
        
        # Dueling DQN architecture
        advantage = self.advantage_stream(global_rep)
        value = self.value_stream(global_rep)
        
        # Combine value and advantage
        q_values = value + (advantage - advantage.mean(dim=1, keepdim=True))
        
        return q_values

class TransformerDQNAgent:
    def __init__(self, env_info, config):
        self.device = config["device"]
        self.num_actions = env_info["num_actions"]
        self.num_event_types = env_info["num_event_types"]
        self.num_colors = env_info["num_colors"]
        self.max_seq_length = config["max_seq_length"]
        
        # Create networks
        self.policy_net = TransformerDQN(
            self.num_event_types,
            self.num_colors,
            config["embed_dim"],
            config["hidden_dim"],
            config["num_heads"],
            config["num_encoder_layers"],
            self.num_actions,
            self.max_seq_length
        ).to(self.device)
        
        self.target_net = copy.deepcopy(self.policy_net)
        self.target_net.eval()  # Target network is in evaluation mode
        
        # Optimizer
        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=config["learning_rate"])
        
        # Replay memory
        self.memory = ReplayMemory(config["memory_size"])
        
        # Training parameters
        self.batch_size = config["batch_size"]
        self.gamma = config["gamma"]
        self.target_update = config["target_update"]
        self.epsilon = config["epsilon_start"]
        self.epsilon_end = config["epsilon_end"]
        self.epsilon_decay = config["epsilon_decay"]
        
        # Tracking
        self.steps_done = 0
        
    def select_action(self, state, eval_mode=False):
        """Select action using epsilon-greedy policy."""
        if eval_mode:
            epsilon = 0.05  # Small epsilon for evaluation
        else:
            epsilon = self.epsilon
            
        # Random action with probability epsilon
        if random.random() < epsilon:
            return random.randrange(self.num_actions)
        
        # Get state representation
        with torch.no_grad():
            # Convert state to tensor and add batch dimension
            state_tensor, mask_tensor = self._prepare_sequence([state])
            state_tensor = state_tensor.to(self.device)
            mask_tensor = mask_tensor.to(self.device)
            
            # Forward pass through network
            q_values = self.policy_net(state_tensor, mask_tensor)
            
            # Return action with highest Q-value
            return q_values.max(1)[1].item()
    
    def _prepare_sequence(self, event_sequences):
        """Convert event sequences to tensor format for the network."""
        batch_size = len(event_sequences)
        max_seq_len = self.max_seq_length
        
        # Initialize tensors
        features_tensor = torch.zeros((batch_size, max_seq_len, 5), dtype=torch.float)
        mask_tensor = torch.ones((batch_size, max_seq_len), dtype=torch.bool)
        
        for b, sequence in enumerate(event_sequences):
            # Process each sequence
            seq_len = min(len(sequence), max_seq_len)
            
            # If sequence is longer than max_seq_len, take most recent events
            events_to_use = sequence[-seq_len:]
            
            for i, event in enumerate(events_to_use):
                # Fill features tensor
                features_tensor[b, i, 0] = event["raw_type"]
                features_tensor[b, i, 1] = event["raw_bg"]
                features_tensor[b, i, 2] = event["raw_fg"]
                features_tensor[b, i, 3] = event["start_time"]
                features_tensor[b, i, 4] = event["end_time"]
                
                # Mark as not masked
                mask_tensor[b, i] = False
        
        return features_tensor, mask_tensor
    
    def optimize_model(self):
        """Training step to optimize the policy network."""
        if len(self.memory) < self.batch_size:
            return
        
        # Sample experiences
        experiences = self.memory.sample(self.batch_size)
        batch = Experience(*zip(*experiences))
        
        # Convert to tensors
        state_batch, mask_batch = self._prepare_sequence(batch.state)
        action_batch = torch.tensor(batch.action, device=self.device)
        reward_batch = torch.tensor(batch.reward, device=self.device, dtype=torch.float)
        
        # For non-terminal next states
        non_final_mask = torch.tensor(
            tuple(map(lambda s: not s, batch.done)),
            device=self.device, dtype=torch.bool
        )
        
        # Filter out terminal states
        non_final_next_states = [s for s, d in zip(batch.next_state, batch.done) if not d]
        
        # Prepare next states tensor
        next_state_batch, next_mask_batch = self._prepare_sequence(non_final_next_states) if non_final_next_states else (None, None)
        
        # Move tensors to device
        state_batch = state_batch.to(self.device)
        mask_batch = mask_batch.to(self.device)
        if next_state_batch is not None:
            next_state_batch = next_state_batch.to(self.device)
            next_mask_batch = next_mask_batch.to(self.device)
        
        # Compute Q(s_t, a)
        state_action_values = self.policy_net(state_batch, mask_batch).gather(1, action_batch.unsqueeze(1))
        
        # Compute max Q(s_{t+1}, a) for all next states
        next_state_values = torch.zeros(self.batch_size, device=self.device)
        if next_state_batch is not None and len(non_final_next_states) > 0:
            with torch.no_grad():
                # Double DQN: use policy_net to select actions and target_net to evaluate them
                next_actions = self.policy_net(next_state_batch, next_mask_batch).max(1)[1].detach()
                next_state_values[non_final_mask] = self.target_net(next_state_batch, next_mask_batch).gather(1, next_actions.unsqueeze(1)).squeeze(1)
        
        # Compute expected Q values
        expected_state_action_values = (next_state_values * self.gamma) + reward_batch
        
        # Compute loss
        criterion = nn.SmoothL1Loss()  # Huber loss for stability
        loss = criterion(state_action_values, expected_state_action_values.unsqueeze(1))
        
        # Optimize
        self.optimizer.zero_grad()
        loss.backward()
        # Gradient clipping to prevent exploding gradients
        torch.nn.utils.clip_grad_norm_(self.policy_net.parameters(), max_norm=1.0)
        self.optimizer.step()
        
        return loss.item()
    
    def update_target_network(self):
        """Update target network by copying policy network weights."""
        self.target_net.load_state_dict(self.policy_net.state_dict())
    
    def update_epsilon(self):
        """Decay epsilon according to schedule."""
        self.epsilon = max(self.epsilon_end, self.epsilon * self.epsilon_decay)
    
    def save_model(self, path):
        """Save model to disk."""
        torch.save({
            'policy_net': self.policy_net.state_dict(),
            'target_net': self.target_net.state_dict(),
            'optimizer': self.optimizer.state_dict(),
            'epsilon': self.epsilon,
            'steps_done': self.steps_done
        }, path)
    
    def load_model(self, path):
        """Load model from disk."""
        checkpoint = torch.load(path)
        self.policy_net.load_state_dict(checkpoint['policy_net'])
        self.target_net.load_state_dict(checkpoint['target_net'])
        self.optimizer.load_state_dict(checkpoint['optimizer'])
        self.epsilon = checkpoint['epsilon']
        self.steps_done = checkpoint['steps_done']

def train_agent(env_name, agent, num_episodes=500, max_steps=1000):
    """Train the agent in the environment."""
    from register_envs import all_types, all_attributes
    
    env = gym.make(env_name)
    rewards = []
    durations = []
    
    # Training loop
    for episode in tqdm(range(num_episodes), desc="Training"):
        # Reset environment
        obs = env.reset()
        done = False
        episode_reward = 0
        steps = 0
        
        # Initialize sequence of events
        events_history = []
        
        # Process first observation
        current_event = event_from_obs_gym(obs, all_types, all_attributes)
        events_history.append(current_event)
        
        # Episode loop
        while not done and steps < max_steps:
            # Select action
            action_idx = agent.select_action(events_history)
            
            # Convert action index to one-hot vector
            action = [0] * env.action_space.n
            action[action_idx] = 1
            
            # Take action
            next_obs, reward, done, _ = env.step(action)
            
            # Process next observation
            next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
            
            # Update history
            next_history = events_history + [next_event]
            
            # Store transition in memory
            agent.memory.push(
                list(events_history),  # Current state
                action_idx,            # Action index
                list(next_history),    # Next state
                reward,                # Reward
                done                   # Done flag
            )
            
            # Move to next state
            events_history = next_history
            episode_reward += reward
            steps += 1
            
            # Optimize model
            loss = agent.optimize_model()
            
            # Update target network
            if steps % agent.target_update == 0:
                agent.update_target_network()
            
            # Update current observation
            obs = next_obs
        
        # Decay epsilon
        agent.update_epsilon()
        
        # Track progress
        rewards.append(episode_reward)
        durations.append(steps)
        
        # Print progress
        if (episode + 1) % 10 == 0:
            avg_reward = sum(rewards[-10:]) / 10
            print(f"Episode {episode+1}/{num_episodes}, Avg Reward: {avg_reward:.2f}, Epsilon: {agent.epsilon:.4f}")
    
    env.close()
    
    # Plot training results
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(rewards)
    plt.title('Training Rewards')
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    
    plt.subplot(1, 2, 2)
    plt.plot(durations)
    plt.title('Episode Durations')
    plt.xlabel('Episode')
    plt.ylabel('Steps')
    
    plt.tight_layout()
    plt.show()
    
    return rewards, durations

def evaluate_agent(env_name, agent, num_episodes=100):
    """Evaluate agent performance."""
    from register_envs import all_types, all_attributes
    
    env = gym.make(env_name)
    rewards = []
    
    for episode in tqdm(range(num_episodes), desc="Evaluating"):
        obs = env.reset()
        done = False
        episode_reward = 0
        
        events_history = []
        
        current_event = event_from_obs_gym(obs, all_types, all_attributes)
        events_history.append(current_event)
        
        # Episode loop
        while not done:
            # Select action (in evaluation mode)
            action_idx = agent.select_action(events_history, eval_mode=True)
            
            # Convert action index to one-hot vector
            action = [0] * env.action_space.n
            action[action_idx] = 1
            
            # Take action
            next_obs, reward, done, _ = env.step(action)
            
            # Process next observation
            next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
            
            # Update history
            events_history.append(next_event)
            episode_reward += reward
            
            # Update current observation
            obs = next_obs
        
        rewards.append(episode_reward)
    
    env.close()
    
    # Print results
    avg_reward = sum(rewards) / len(rewards)
    print(f"Average reward over {num_episodes} episodes: {avg_reward:.2f}")
    
    # Plot results
    plt.figure(figsize=(8, 4))
    plt.plot(rewards)
    plt.axhline(avg_reward, color='r', linestyle='--', label=f'Avg: {avg_reward:.2f}')
    plt.title('Evaluation Rewards')
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.legend()
    plt.show()
    
    return rewards

def main():
    from register_envs import all_types, all_attributes, register_custom_envs
    register_custom_envs()
    
    env_name = "OpenTheChests-v2"
    print(f"Using environment: {env_name}")
    
    # Get environment information
    env = gym.make(env_name)
    num_actions = env.action_space.n
    
    # Get information about event types and attributes
    num_event_types = len(all_types)
    num_colors = len(all_attributes["bg"])  # Assuming bg and fg have same number of colors
    
    env_info = {
        "num_actions": num_actions,
        "num_event_types": num_event_types,
        "num_colors": num_colors
    }
    
    print(f"Environment info: {env_info}")
    env.close()
    
    # Create agent
    agent = TransformerDQNAgent(env_info, CONFIG)
    
    # Train agent
    print("Starting training...")
    rewards, durations = train_agent(env_name, agent, num_episodes=500)
    
    # Save trained agent
    agent.save_model("transformer_dqn_agent.pt")
    print("Model saved as 'transformer_dqn_agent.pt'")
    
    # Evaluate agent
    print("Evaluating agent...")
    eval_rewards = evaluate_agent(env_name, agent)
    
    # Print final results
    avg_reward = sum(eval_rewards) / len(eval_rewards)
    print(f"Final average reward: {avg_reward:.2f}")
    
    return avg_reward

if __name__ == "__main__":
    main()
````

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import gym
import random
from collections import deque, namedtuple
import matplotlib.pyplot as plt
from tqdm import tqdm
import copy
import os # Pour la gestion des chemins de sauvegarde

# --- Configuration ---
CONFIG = {
    "env_name": "OpenTheChests-v2", # Nom de l'environnement Gym
    "batch_size": 256,
    "learning_rate": 0.008, # Taux d'apprentissage pour Adam
    "gamma": 0.99,           # Facteur d'actualisation des rÃ©compenses futures
    "target_update": 100,    # FrÃ©quence de mise Ã  jour du rÃ©seau cible (en Ã©tapes)
    "memory_size": 50000,    # Taille du buffer de rejeu (augmentÃ©e)
    "hidden_dim": 256,       # Dimension cachÃ©e dans le Transformer et le MLP
    "embed_dim": 128,        # Dimension des embeddings d'Ã©vÃ©nements
    "num_heads": 8,          # Nombre de tÃªtes d'attention dans le Transformer
    "num_encoder_layers": 4, # Nombre de couches dans l'encodeur Transformer (rÃ©duit pour potentiellement accÃ©lÃ©rer/stabiliser)
    "max_seq_length": 50,    # Longueur maximale de la sÃ©quence d'Ã©vÃ©nements
    "epsilon_start": 1.0,    # Valeur initiale d'epsilon (exploration)
    "epsilon_end": 0.05,     # Valeur minimale d'epsilon
    "epsilon_decay": 0.999,  # Taux de dÃ©croissance d'epsilon (plus lent pour plus d'exploration)
    "gradient_clip_norm": 1.0, # Norme maximale pour le clipping de gradient
    "num_episodes": 5,    # Nombre d'Ã©pisodes pour l'entraÃ®nement (augmentÃ©)
    "max_steps_per_episode": 500, # Nombre maximum de pas par Ã©pisode
    "save_path": "transformer_dqn_agent.pt", # Chemin pour sauvegarder le modÃ¨le
    "device": torch.device("cuda" if torch.cuda.is_available() else "cpu"), # Utilisation GPU si disponible
    "padding_idx": 0 # Index utilisÃ© pour le padding
}

print(f"Using device: {CONFIG['device']}")

# --- Traitement des Observations ---
# Assurez-vous que register_envs est importÃ© et exÃ©cutÃ© si nÃ©cessaire
try:
    from register_envs import all_types, all_attributes, register_custom_envs
    register_custom_envs()
    print("Custom environments registered.")
except ImportError:
    print("Warning: Could not import or register custom environments from 'register_envs'.")
    # DÃ©finir des valeurs par dÃ©faut si l'import Ã©choue (Ã  adapter si nÃ©cessaire)
    all_types = ["default_type"]
    all_attributes = {"bg": ["default_color"], "fg": ["default_color"]}

def event_from_obs_gym(obs, all_types, all_attributes):
    """Extrait les informations d'Ã©vÃ©nement d'une observation Gym.
    GÃ¨re les dictionnaires et les listes/tableaux numpy.
    Retourne un dictionnaire structurÃ© avec les caractÃ©ristiques brutes et interprÃ©tÃ©es.
    """
    event = {
        "raw_type": CONFIG['padding_idx'], # Utilise l'index de padding par dÃ©faut
        "raw_bg": CONFIG['padding_idx'],
        "raw_fg": CONFIG['padding_idx'],
        "start_time": 0.0,
        "end_time": 0.0,
        "symbol": "pad",
        "bg_color": "pad",
        "symbol_color": "pad"
    }
    try:
        num_types = len(all_types)
        num_bg_colors = len(all_attributes.get("bg", []))
        num_fg_colors = len(all_attributes.get("fg", []))

        if isinstance(obs, dict):
            raw_type = obs.get("e_type", CONFIG['padding_idx'])
            raw_bg = obs.get("bg", CONFIG['padding_idx'])
            raw_fg = obs.get("fg", CONFIG['padding_idx'])
            start_time = obs.get("start", [0.0])[0] if isinstance(obs.get("start"), (list, np.ndarray)) and len(obs["start"]) > 0 else 0.0
            end_time = obs.get("end", [0.0])[0] if isinstance(obs.get("end"), (list, np.ndarray)) and len(obs["end"]) > 0 else 0.0

        elif isinstance(obs, (list, np.ndarray)):
            # HypothÃ¨se de structure : [e_type, bg, fg, start_time, end_time]
            # Ajustez les indices si la structure rÃ©elle est diffÃ©rente
            raw_type = int(obs[0]) if len(obs) > 0 else CONFIG['padding_idx']
            raw_bg = int(obs[1]) if len(obs) > 1 else CONFIG['padding_idx']
            raw_fg = int(obs[2]) if len(obs) > 2 else CONFIG['padding_idx']
            start_time = float(obs[3]) if len(obs) > 3 else 0.0
            end_time = float(obs[4]) if len(obs) > 4 else 0.0
        else:
            print(f"Warning: Unexpected observation type: {type(obs)}. Using default event.")
            return event # Retourne l'Ã©vÃ©nement de padding par dÃ©faut

        # VÃ©rification des indices avant d'accÃ©der aux listes
        event["raw_type"] = raw_type
        event["raw_bg"] = raw_bg
        event["raw_fg"] = raw_fg
        event["start_time"] = float(start_time)
        event["end_time"] = float(end_time)
        event["symbol"] = all_types[raw_type] if 0 <= raw_type < num_types else "unknown"
        event["bg_color"] = all_attributes["bg"][raw_bg] if 0 <= raw_bg < num_bg_colors else "unknown"
        event["symbol_color"] = all_attributes["fg"][raw_fg] if 0 <= raw_fg < num_fg_colors else "unknown"

    except IndexError as e:
        print(f"Error processing observation (IndexError): {e}")
        print(f"Observation: {obs}")
        print(f"Indices: type={event.get('raw_type', 'N/A')}, bg={event.get('raw_bg', 'N/A')}, fg={event.get('raw_fg', 'N/A')}")
        print(f"List lengths: types={len(all_types)}, bg={len(all_attributes.get('bg',[]))}, fg={len(all_attributes.get('fg',[]))}")
        # Retourne l'Ã©vÃ©nement partiellement rempli ou un Ã©vÃ©nement par dÃ©faut
    except Exception as e:
        print(f"Error processing observation: {e}")
        print(f"Observation: {obs}")
        # Retourne l'Ã©vÃ©nement de padding par dÃ©faut en cas d'erreur grave
        event = {k: CONFIG['padding_idx'] if 'raw' in k else (0.0 if 'time' in k else 'error') for k in event}


    # Assurez-vous que les types bruts sont des entiers pour l'embedding
    event["raw_type"] = int(event["raw_type"])
    event["raw_bg"] = int(event["raw_bg"])
    event["raw_fg"] = int(event["raw_fg"])

    return event


# --- Buffer de Rejeu ---
Experience = namedtuple('Experience', ('state', 'action', 'next_state', 'reward', 'done'))

class ReplayMemory:
    """Buffer de rejeu de taille fixe pour stocker les expÃ©riences."""
    def __init__(self, capacity):
        self.memory = deque(maxlen=capacity)

    def push(self, *args):
        """Ajoute une expÃ©rience au buffer."""
        self.memory.append(Experience(*args))

    def sample(self, batch_size):
        """Ã‰chantillonne un batch d'expÃ©riences alÃ©atoirement."""
        return random.sample(self.memory, batch_size)

    def __len__(self):
        """Retourne la taille actuelle du buffer."""
        return len(self.memory)

# --- Composants du RÃ©seau ---
class PositionalEncoding(nn.Module):
    """Ajoute l'encodage positionnel aux embeddings d'entrÃ©e."""
    def __init__(self, d_model, max_seq_length=100):
        super().__init__()
        position = torch.arange(max_seq_length).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
        pe = torch.zeros(1, max_seq_length, d_model)
        pe[0, :, 0::2] = torch.sin(position * div_term)
        pe[0, :, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe) # Enregistre pe comme buffer, pas comme paramÃ¨tre

    def forward(self, x):
        """ x: Tensor, shape [batch_size, seq_len, embedding_dim] """
        x = x + self.pe[:, :x.size(1)]
        return x

class EventEncoder(nn.Module):
    """Encode les caractÃ©ristiques d'un Ã©vÃ©nement individuel en un embedding."""
    def __init__(self, num_event_types, num_colors, embed_dim, padding_idx):
        super().__init__()
        # +1 pour gÃ©rer l'index de padding (0)
        # padding_idx=padding_idx indique Ã  la couche d'ignorer cet index pendant l'entraÃ®nement
        # et de sortir un vecteur nul.
        self.type_embedding = nn.Embedding(num_event_types + 1, embed_dim // 3, padding_idx=padding_idx)
        self.bg_embedding = nn.Embedding(num_colors + 1, embed_dim // 3, padding_idx=padding_idx)
        self.fg_embedding = nn.Embedding(num_colors + 1, embed_dim // 3, padding_idx=padding_idx)

        # Encodeur simple pour les caractÃ©ristiques temporelles (start_time, end_time)
        # On pourrait utiliser un encodeur plus complexe si le temps est crucial
        self.time_encoder = nn.Sequential(
            nn.Linear(2, embed_dim // 2), # RÃ©duit la dimension pour Ã©quilibrer
            nn.ReLU(),
            nn.Linear(embed_dim // 2, embed_dim // 3) # Ajuste pour correspondre aux autres embeddings
        )

        # Projection finale pour combiner les embeddings et atteindre embed_dim
        # La dimension d'entrÃ©e est la somme des dimensions des embeddings partiels
        total_embedding_dim = (embed_dim // 3) * 4 # 3 pour type/bg/fg + 1 pour time
        self.proj = nn.Linear(total_embedding_dim, embed_dim)
        self.norm = nn.LayerNorm(embed_dim)
        self.relu = nn.ReLU()

    def forward(self, event_features):
        """
        event_features: Tensor, shape [batch_size, seq_len, num_features]
                       num_features = 5 (raw_type, raw_bg, raw_fg, start_time, end_time)
        """
        # Extraction des caractÃ©ristiques
        # Assurez-vous que les indices sont de type Long pour les embeddings
        event_type = event_features[:, :, 0].long()
        bg_color = event_features[:, :, 1].long()
        fg_color = event_features[:, :, 2].long()
        time_features = event_features[:, :, 3:5].float() # start_time, end_time

        # Calcul des embeddings
        type_emb = self.type_embedding(event_type)
        bg_emb = self.bg_embedding(bg_color)
        fg_emb = self.fg_embedding(fg_color)
        time_emb = self.time_encoder(time_features)

        # ConcatÃ©nation des embeddings
        combined = torch.cat([type_emb, bg_emb, fg_emb, time_emb], dim=2)

        # Projection et normalisation
        output = self.proj(combined)
        output = self.norm(output)
        output = self.relu(output) # Ajout d'une activation
        return output

class TransformerDQN(nn.Module):
    """RÃ©seau DQN basÃ© sur un Transformer avec architecture Dueling."""
    def __init__(self, num_event_types, num_colors, embed_dim, hidden_dim,
                 num_heads, num_encoder_layers, num_actions, max_seq_length, padding_idx):
        super().__init__()
        self.padding_idx = padding_idx

        # Encodeur d'Ã©vÃ©nements
        self.event_encoder = EventEncoder(num_event_types, num_colors, embed_dim, padding_idx)

        # Encodage positionnel
        self.pos_encoder = PositionalEncoding(embed_dim, max_seq_length)

        # Couche d'encodeur Transformer standard
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=num_heads,
            dim_feedforward=hidden_dim,
            dropout=0.1, # Dropout pour la rÃ©gularisation
            activation='relu', # Utilisation de ReLU
            batch_first=True # Important: les donnÃ©es sont [batch, seq, feature]
        )
        # Normalisation avant l'encodeur Transformer
        self.encoder_norm = nn.LayerNorm(embed_dim)
        self.transformer_encoder = nn.TransformerEncoder(
            encoder_layer,
            num_layers=num_encoder_layers,
            norm=nn.LayerNorm(embed_dim) # Normalisation aprÃ¨s l'encodeur
        )

        # Architecture Dueling DQN
        # SÃ©pare l'estimation de la valeur de l'Ã©tat (V) et de l'avantage de chaque action (A)
        # Avantage Stream
        self.advantage_stream = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, num_actions)
        )

        # Value Stream
        self.value_stream = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, 1) # Sortie unique pour la valeur de l'Ã©tat
        )

    def forward(self, x, src_key_padding_mask=None):
        """
        x: Tensor, shape [batch_size, seq_len, num_features]
           num_features = 5 (raw_type, raw_bg, raw_fg, start_time, end_time)
        src_key_padding_mask: Tensor boolÃ©en, shape [batch_size, seq_len]
                              True indique une position paddÃ©e Ã  ignorer par l'attention.
        """
        # 1. Encoder les Ã©vÃ©nements individuels
        # shape: [batch_size, seq_len, embed_dim]
        event_embeddings = self.event_encoder(x)

        # 2. Ajouter l'encodage positionnel
        # shape: [batch_size, seq_len, embed_dim]
        pos_encoded_embeddings = self.pos_encoder(event_embeddings)
        pos_encoded_embeddings = self.encoder_norm(pos_encoded_embeddings) # Normalisation avant Transformer

        # 3. Passer Ã  travers l'encodeur Transformer
        # Le masque permet d'ignorer les Ã©lÃ©ments de padding dans les calculs d'attention
        # shape: [batch_size, seq_len, embed_dim]
        transformer_output = self.transformer_encoder(
            pos_encoded_embeddings,
            src_key_padding_mask=src_key_padding_mask
        )

        # 4. AgrÃ©ger les sorties du Transformer
        # Utiliser la sortie du premier token non paddÃ© (si on utilisait un token [CLS])
        # ou moyenner les sorties non paddÃ©es. Ici, on moyenne sur la dimension de sÃ©quence.
        # Il faut masquer la moyenne pour ne pas inclure les Ã©lÃ©ments paddÃ©s.
        if src_key_padding_mask is not None:
            # Inverser le masque pour le calcul de la moyenne (True = garder, False = ignorer)
            mask_for_mean = ~src_key_padding_mask.unsqueeze(-1).expand_as(transformer_output)
            masked_output = transformer_output * mask_for_mean
            # Compter les Ã©lÃ©ments non paddÃ©s par sÃ©quence pour normaliser la somme
            non_padded_count = mask_for_mean.sum(dim=1)
            # Ã‰viter la division par zÃ©ro si une sÃ©quence est entiÃ¨rement paddÃ©e (ne devrait pas arriver avec une bonne prÃ©paration)
            non_padded_count = torch.max(non_padded_count, torch.ones_like(non_padded_count))
            aggregated_output = masked_output.sum(dim=1) / non_padded_count
        else:
            # Si pas de masque, simple moyenne (moins robuste si padding prÃ©sent)
            aggregated_output = transformer_output.mean(dim=1)
        # shape: [batch_size, embed_dim]

        # 5. Calculer V(s) et A(s, a) via les streams Dueling
        value = self.value_stream(aggregated_output)          # shape: [batch_size, 1]
        advantages = self.advantage_stream(aggregated_output) # shape: [batch_size, num_actions]

        # 6. Combiner V et A pour obtenir les Q-values
        # Q(s, a) = V(s) + (A(s, a) - mean(A(s, a')))
        q_values = value + (advantages - advantages.mean(dim=1, keepdim=True))
        # shape: [batch_size, num_actions]

        return q_values

# --- Agent DQN ---
class TransformerDQNAgent:
    """Agent utilisant le TransformerDQN pour apprendre."""
    def __init__(self, env_info, config):
        self.device = config["device"]
        self.num_actions = env_info["num_actions"]
        self.num_event_types = env_info["num_event_types"]
        self.num_colors = env_info["num_colors"]
        self.max_seq_length = config["max_seq_length"]
        self.padding_idx = config["padding_idx"]
        self.config = config # Stocker la config pour accÃ¨s facile

        # Initialisation des rÃ©seaux (politique et cible)
        self.policy_net = TransformerDQN(
            self.num_event_types, self.num_colors,
            config["embed_dim"], config["hidden_dim"], config["num_heads"],
            config["num_encoder_layers"], self.num_actions, self.max_seq_length,
            self.padding_idx
        ).to(self.device)

        self.target_net = TransformerDQN(
            self.num_event_types, self.num_colors,
            config["embed_dim"], config["hidden_dim"], config["num_heads"],
            config["num_encoder_layers"], self.num_actions, self.max_seq_length,
            self.padding_idx
        ).to(self.device)
        # Copier les poids du rÃ©seau de politique vers le rÃ©seau cible
        self.target_net.load_state_dict(self.policy_net.state_dict())
        self.target_net.eval() # Mettre le rÃ©seau cible en mode Ã©valuation (pas de dropout, etc.)

        # Optimiseur Adam
        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=config["learning_rate"])

        # Buffer de rejeu
        self.memory = ReplayMemory(config["memory_size"])

        # ParamÃ¨tres d'entraÃ®nement
        self.batch_size = config["batch_size"]
        self.gamma = config["gamma"]
        self.target_update_freq = config["target_update"] # RenommÃ© pour clartÃ©
        self.epsilon = config["epsilon_start"]
        self.epsilon_end = config["epsilon_end"]
        self.epsilon_decay = config["epsilon_decay"]
        self.gradient_clip_norm = config["gradient_clip_norm"]

        # Suivi de l'entraÃ®nement
        self.total_steps = 0 # Compteur global des Ã©tapes

    def _prepare_sequence(self, event_sequences_batch):
        """Convertit un batch de sÃ©quences d'Ã©vÃ©nements en tenseurs paddÃ©s et crÃ©e le masque."""
        batch_size = len(event_sequences_batch)
        max_len = self.max_seq_length

        # Initialiser les tenseurs avec des zÃ©ros (qui correspondent au padding_idx pour les features)
        # Features: type, bg, fg, start_time, end_time
        features_tensor = torch.full((batch_size, max_len, 5), float(self.padding_idx), dtype=torch.float)
        # Masque: True pour les positions paddÃ©es, False sinon
        mask_tensor = torch.ones((batch_size, max_len), dtype=torch.bool)

        for b_idx, sequence in enumerate(event_sequences_batch):
            # Tronquer la sÃ©quence si elle est trop longue, en gardant les Ã©vÃ©nements les plus rÃ©cents
            seq_len = min(len(sequence), max_len)
            events_to_use = sequence[-seq_len:]

            for i, event in enumerate(events_to_use):
                # Remplir le tenseur de caractÃ©ristiques
                # L'indice dans le tenseur est `max_len - seq_len + i` pour aligner Ã  droite (plus rÃ©cent)
                # Ou simplement `i` si on aligne Ã  gauche (plus simple)
                tensor_idx = i # Alignement Ã  gauche
                features_tensor[b_idx, tensor_idx, 0] = float(event["raw_type"])
                features_tensor[b_idx, tensor_idx, 1] = float(event["raw_bg"])
                features_tensor[b_idx, tensor_idx, 2] = float(event["raw_fg"])
                features_tensor[b_idx, tensor_idx, 3] = float(event["start_time"])
                features_tensor[b_idx, tensor_idx, 4] = float(event["end_time"])

                # Mettre le masque Ã  False pour cette position (non paddÃ©e)
                mask_tensor[b_idx, tensor_idx] = False

        return features_tensor, mask_tensor

    def select_action(self, state_sequence, eval_mode=False):
        """SÃ©lectionne une action en utilisant la politique epsilon-greedy."""
        current_epsilon = 0.01 if eval_mode else self.epsilon # Utiliser un petit epsilon fixe en Ã©valuation

        # Exploration: choisir une action alÃ©atoire
        if random.random() < current_epsilon:
            return random.randrange(self.num_actions)
        # Exploitation: choisir la meilleure action selon le rÃ©seau de politique
        else:
            with torch.no_grad(): # Pas besoin de calculer les gradients ici
                # PrÃ©parer la sÃ©quence d'Ã©tat pour le rÃ©seau (ajouter une dimension batch)
                state_tensor, mask_tensor = self._prepare_sequence([state_sequence])
                # DÃ©placer les tenseurs vers le bon device
                state_tensor = state_tensor.to(self.device)
                mask_tensor = mask_tensor.to(self.device)

                # Obtenir les Q-values du rÃ©seau de politique
                self.policy_net.eval() # Passer en mode Ã©valuation temporairement si nÃ©cessaire
                q_values = self.policy_net(state_tensor, mask_tensor)
                self.policy_net.train() # Revenir en mode entraÃ®nement

                # Choisir l'action avec la Q-value la plus Ã©levÃ©e
                action = q_values.max(1)[1].item() # .item() pour obtenir la valeur Python
                return action

    def optimize_model(self):
        """Effectue une Ã©tape d'optimisation du rÃ©seau de politique."""
        # Attendre d'avoir assez d'expÃ©riences dans le buffer
        if len(self.memory) < self.batch_size:
            return None # Pas assez d'Ã©chantillons

        # Ã‰chantillonner un batch d'expÃ©riences
        experiences = self.memory.sample(self.batch_size)
        # Transposer le batch (convertit un batch d'Experiences en une Experience de batches)
        batch = Experience(*zip(*experiences))

        # PrÃ©parer les tenseurs pour les Ã©tats courants et suivants
        state_batch_tensor, state_mask_tensor = self._prepare_sequence(batch.state)
        next_state_batch_tensor, next_state_mask_tensor = self._prepare_sequence(batch.next_state)

        # Convertir les actions, rÃ©compenses et 'done' en tenseurs
        # Assurez-vous qu'ils sont sur le bon device
        action_batch = torch.tensor(batch.action, device=self.device, dtype=torch.long).unsqueeze(1) # [batch_size, 1]
        reward_batch = torch.tensor(batch.reward, device=self.device, dtype=torch.float).unsqueeze(1) # [batch_size, 1]
        # Masque pour les Ã©tats non finaux (done=False)
        non_final_mask = torch.tensor(tuple(map(lambda d: not d, batch.done)), device=self.device, dtype=torch.bool)

        # DÃ©placer les tenseurs d'Ã©tat/masque vers le device
        state_batch_tensor = state_batch_tensor.to(self.device)
        state_mask_tensor = state_mask_tensor.to(self.device)
        next_state_batch_tensor = next_state_batch_tensor.to(self.device)
        next_state_mask_tensor = next_state_mask_tensor.to(self.device)

        # --- Calcul de la perte ---
        # 1. Calculer Q(s_t, a_t) pour les actions rÃ©ellement prises
        # Le rÃ©seau de politique est utilisÃ© ici.
        # gather(1, action_batch) sÃ©lectionne la Q-value correspondant Ã  l'action prise.
        current_q_values = self.policy_net(state_batch_tensor, state_mask_tensor).gather(1, action_batch)

        # 2. Calculer V(s_{t+1}) pour les Ã©tats suivants
        # Utilisation de Double DQN:
        # - Le rÃ©seau de POLITIQUE choisit la MEILLEURE action pour l'Ã©tat suivant: a' = argmax_a Q_policy(s_{t+1}, a)
        # - Le rÃ©seau CIBLE Ã©value cette action: V(s_{t+1}) = Q_target(s_{t+1}, a')
        next_q_values_target = torch.zeros(self.batch_size, 1, device=self.device) # Initialiser Ã  zÃ©ro (pour les Ã©tats terminaux)
        if torch.any(non_final_mask): # S'il y a au moins un Ã©tat non final
            # SÃ©lectionner les Ã©tats suivants non finaux et leurs masques
            non_final_next_states = next_state_batch_tensor[non_final_mask]
            non_final_next_masks = next_state_mask_tensor[non_final_mask]

            # Obtenir les Q-values du rÃ©seau de politique pour choisir les actions
            with torch.no_grad(): # Pas de gradient pour la sÃ©lection d'action
                 # Passer policy_net en eval temporairement pour la sÃ©lection d'action
                self.policy_net.eval()
                next_state_q_values_policy = self.policy_net(non_final_next_states, non_final_next_masks)
                self.policy_net.train() # Remettre en mode train

                # Choisir les meilleures actions selon le rÃ©seau de politique
                best_next_actions = next_state_q_values_policy.max(1)[1].unsqueeze(1) # [num_non_final, 1]

                # Ã‰valuer ces actions avec le rÃ©seau cible (qui est en mode eval)
                target_q_values = self.target_net(non_final_next_states, non_final_next_masks).gather(1, best_next_actions)

            # Mettre Ã  jour les valeurs pour les Ã©tats non finaux
            next_q_values_target[non_final_mask] = target_q_values.detach() # detach() car c'est une cible

        # 3. Calculer la valeur Q cible (Bellman equation)
        # target = r + gamma * V(s_{t+1})
        expected_q_values = reward_batch + (self.gamma * next_q_values_target)

        # 4. Calculer la perte (Huber Loss / Smooth L1 Loss)
        criterion = nn.SmoothL1Loss()
        loss = criterion(current_q_values, expected_q_values)

        # --- Optimisation ---
        # RÃ©initialiser les gradients
        self.optimizer.zero_grad()
        # RÃ©tropropagation
        loss.backward()
        # Clipper les gradients pour Ã©viter l'explosion (stabilise l'entraÃ®nement)
        torch.nn.utils.clip_grad_norm_(self.policy_net.parameters(), self.gradient_clip_norm)
        # Mettre Ã  jour les poids du rÃ©seau de politique
        self.optimizer.step()

        return loss.item() # Retourner la valeur de la perte

    def update_target_network(self):
        """Copie les poids du rÃ©seau de politique vers le rÃ©seau cible."""
        print(f"Updating target network at step {self.total_steps}")
        self.target_net.load_state_dict(self.policy_net.state_dict())

    def update_epsilon(self):
        """RÃ©duit epsilon linÃ©airement ou exponentiellement."""
        self.epsilon = max(self.epsilon_end, self.epsilon * self.epsilon_decay)

    def step_update(self):
        """Met Ã  jour l'Ã©tat de l'agent aprÃ¨s chaque pas dans l'environnement."""
        self.total_steps += 1
        self.update_epsilon() # DÃ©croissance d'epsilon Ã  chaque pas

        # Mettre Ã  jour le rÃ©seau cible pÃ©riodiquement
        if self.total_steps % self.target_update_freq == 0:
            self.update_target_network()

    def save_model(self, path):
        """Sauvegarde l'Ã©tat de l'agent (rÃ©seaux, optimiseur, epsilon)."""
        print(f"Saving model to {path}...")
        torch.save({
            'policy_net_state_dict': self.policy_net.state_dict(),
            'target_net_state_dict': self.target_net.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict(),
            'epsilon': self.epsilon,
            'total_steps': self.total_steps,
            'config': self.config # Sauvegarder aussi la config peut Ãªtre utile
        }, path)

    def load_model(self, path):
        """Charge l'Ã©tat de l'agent depuis un fichier."""
        if os.path.exists(path):
            print(f"Loading model from {path}...")
            checkpoint = torch.load(path, map_location=self.device) # Charger sur le bon device
            self.policy_net.load_state_dict(checkpoint['policy_net_state_dict'])
            self.target_net.load_state_dict(checkpoint['target_net_state_dict'])
            self.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            self.epsilon = checkpoint['epsilon']
            self.total_steps = checkpoint['total_steps']
            # Assurer que les rÃ©seaux sont sur le bon device aprÃ¨s chargement
            self.policy_net.to(self.device)
            self.target_net.to(self.device)
            # Mettre target_net en mode Ã©valuation
            self.target_net.eval()
            print("Model loaded successfully.")
        else:
            print(f"Warning: Model file not found at {path}. Starting from scratch.")

# --- Fonctions d'EntraÃ®nement et d'Ã‰valuation ---

def train_agent(agent, config):
    """EntraÃ®ne l'agent dans l'environnement spÃ©cifiÃ©."""
    env_name = config["env_name"]
    num_episodes = config["num_episodes"]
    max_steps_per_episode = config["max_steps_per_episode"]

    env = gym.make(env_name)
    print(f"Environment '{env_name}' created.")
    print(f"Action space: {env.action_space}")
    print(f"Observation space: {env.observation_space}")

    rewards_history = []
    durations_history = []
    losses_history = [] # Pour suivre la perte

    # Boucle d'entraÃ®nement sur les Ã©pisodes
    for episode in tqdm(range(num_episodes), desc="Training Progress"):
        obs = env.reset()
        done = False
        episode_reward = 0
        episode_steps = 0
        episode_losses = []

        # Initialiser l'historique des Ã©vÃ©nements pour cet Ã©pisode
        # Commencer avec l'Ã©vÃ©nement initial
        current_event = event_from_obs_gym(obs, all_types, all_attributes)
        # GÃ©rer le cas oÃ¹ l'observation initiale est invalide
        if current_event is None or current_event['symbol'] == 'error':
             print(f"Error processing initial observation: {obs}. Skipping episode {episode+1}.")
             continue # Passer Ã  l'Ã©pisode suivant

        events_history = deque([current_event], maxlen=agent.max_seq_length) # Utiliser deque pour gÃ©rer la taille

        # Boucle de l'Ã©pisode
        while not done and episode_steps < max_steps_per_episode:
            # 1. SÃ©lectionner une action
            # Convertir deque en liste pour select_action
            action_idx = agent.select_action(list(events_history))
            action = np.zeros(agent.num_actions, dtype=int)
            action[action_idx] = 1 # Convertir l'index d'action en one-hot

            # 2. ExÃ©cuter l'action dans l'environnement
            # CORRECTION: Passer l'index de l'action (entier) directement
            next_obs, reward, done, info = env.step(action)

            # 3. Traiter la nouvelle observation
            next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
            if next_event is None or next_event['symbol'] == 'error':
                print(f"Error processing observation: {next_obs} at step {episode_steps}. Ending episode.")
                done = True # Terminer l'Ã©pisode si l'observation est invalide
                next_history_list = list(events_history) # Utiliser l'Ã©tat courant comme Ã©tat suivant fictif
            else:
                 # CrÃ©er l'historique suivant (ne pas modifier l'original avant push)
                next_history = events_history.copy()
                next_history.append(next_event)
                next_history_list = list(next_history)


            # 4. Stocker l'expÃ©rience dans le buffer de rejeu
            # Utiliser list(events_history) pour obtenir une copie immuable de l'Ã©tat
            agent.memory.push(
                list(events_history), # Ã‰tat courant (liste d'Ã©vÃ©nements)
                action_idx,           # Action prise (index)
                next_history_list,    # Ã‰tat suivant (liste d'Ã©vÃ©nements)
                reward,               # RÃ©compense obtenue
                done                  # Indicateur de fin d'Ã©pisode
            )

            # 5. Mettre Ã  jour l'historique des Ã©vÃ©nements pour le pas suivant
            if not (next_event is None or next_event['symbol'] == 'error'):
                 events_history.append(next_event) # Ajoute le nouvel Ã©vÃ©nement (deque gÃ¨re la taille max)


            # 6. Optimiser le modÃ¨le (si assez d'expÃ©riences)
            loss = agent.optimize_model()
            if loss is not None:
                episode_losses.append(loss)

            # 7. Mettre Ã  jour l'Ã©tat de l'agent (compteurs, epsilon, rÃ©seau cible)
            agent.step_update()

            # Mise Ã  jour pour le prochain pas
            episode_reward += reward
            episode_steps += 1
            # obs = next_obs # Pas nÃ©cessaire car on utilise l'historique

        # Fin de l'Ã©pisode
        rewards_history.append(episode_reward)
        durations_history.append(episode_steps)
        if episode_losses:
            losses_history.append(np.mean(episode_losses))

        # Affichage pÃ©riodique des progrÃ¨s
        if (episode + 1) % 50 == 0:
            avg_reward = np.mean(rewards_history[-50:])
            avg_duration = np.mean(durations_history[-50:])
            avg_loss = np.mean(losses_history[-50:]) if losses_history else 0
            print(f"\nEpisode {episode+1}/{num_episodes} | Avg Reward (last 50): {avg_reward:.2f} | "
                  f"Avg Duration: {avg_duration:.1f} | Avg Loss: {avg_loss:.4f} | Epsilon: {agent.epsilon:.4f}")
            # Sauvegarde intermÃ©diaire du modÃ¨le
            agent.save_model(config["save_path"] + f"_ep{episode+1}")


    env.close()
    print("Training finished.")

    # Affichage des courbes d'entraÃ®nement
    plt.figure(figsize=(18, 5))

    plt.subplot(1, 3, 1)
    plt.plot(rewards_history)
    plt.title('RÃ©compense par Ã©pisode')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('RÃ©compense totale')
    # Ajouter une moyenne mobile pour lisser la courbe
    if len(rewards_history) >= 50:
        moving_avg = np.convolve(rewards_history, np.ones(50)/50, mode='valid')
        plt.plot(np.arange(49, len(rewards_history)), moving_avg, label='Moyenne mobile (50 Ã©p.)', color='red')
        plt.legend()


    plt.subplot(1, 3, 2)
    plt.plot(durations_history)
    plt.title('DurÃ©e des Ã©pisodes')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('Nombre de pas')
    if len(durations_history) >= 50:
        moving_avg_dur = np.convolve(durations_history, np.ones(50)/50, mode='valid')
        plt.plot(np.arange(49, len(durations_history)), moving_avg_dur, label='Moyenne mobile (50 Ã©p.)', color='red')
        plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(losses_history)
    plt.title('Perte moyenne par Ã©pisode')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('Perte (SmoothL1Loss)')
    if len(losses_history) >= 50:
        moving_avg_loss = np.convolve(losses_history, np.ones(50)/50, mode='valid')
        plt.plot(np.arange(49, len(losses_history)), moving_avg_loss, label='Moyenne mobile (50 Ã©p.)', color='red')
        plt.legend()


    plt.tight_layout()
    plt.savefig("training_curves.png") # Sauvegarder le graphique
    plt.show()

    return rewards_history, durations_history

def evaluate_agent(agent, config, num_episodes=50):
    """Ã‰value les performances de l'agent entraÃ®nÃ©."""
    env_name = config["env_name"]
    max_steps_per_episode = config["max_steps_per_episode"]

    env = gym.make(env_name)
    rewards = []
    print(f"\nStarting evaluation for {num_episodes} episodes...")

    for episode in tqdm(range(num_episodes), desc="Evaluating"):
        obs = env.reset()
        done = False
        episode_reward = 0
        steps = 0

        current_event = event_from_obs_gym(obs, all_types, all_attributes)
        if current_event is None or current_event['symbol'] == 'error':
             print(f"Error processing initial observation during evaluation. Skipping episode {episode+1}.")
             continue
        events_history = deque([current_event], maxlen=agent.max_seq_length)

        while not done and steps < max_steps_per_episode:
            # SÃ©lectionner l'action en mode Ã©valuation (pas d'epsilon-greedy alÃ©atoire)
            action_idx = agent.select_action(list(events_history), eval_mode=True)
            action = np.zeros(agent.num_actions, dtype=int)
            action[action_idx] = 1 # Convertir l'index d'action en one-hot

            # ExÃ©cuter l'action
            next_obs, reward, done, _ = env.step(action)

            # Traiter la nouvelle observation
            next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
            if next_event is None or next_event['symbol'] == 'error':
                print(f"Error processing observation during evaluation step {steps}. Ending episode.")
                done = True
            else:
                events_history.append(next_event)

            episode_reward += reward
            steps += 1
            # obs = next_obs # Pas nÃ©cessaire

        rewards.append(episode_reward)

    env.close()

    # Afficher les rÃ©sultats
    avg_reward = np.mean(rewards)
    std_reward = np.std(rewards)
    print(f"\nEvaluation finished.")
    print(f"Average reward over {num_episodes} episodes: {avg_reward:.2f} +/- {std_reward:.2f}")

    # Afficher le graphique des rÃ©compenses d'Ã©valuation
    plt.figure(figsize=(10, 5))
    plt.plot(rewards, label='RÃ©compense par Ã©pisode')
    plt.axhline(avg_reward, color='r', linestyle='--', label=f'Moyenne: {avg_reward:.2f}')
    plt.title('RÃ©compenses en Ã‰valuation')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('RÃ©compense totale')
    plt.legend()
    plt.savefig("evaluation_rewards.png")
    plt.show()

    return rewards

# --- Fonction Principale ---
def main():
    # Assurez-vous que les environnements personnalisÃ©s sont enregistrÃ©s
    try:
        register_custom_envs()
    except NameError:
        print("`register_custom_envs` not found or failed. Using standard Gym environments.")

    env_name = CONFIG["env_name"]
    print(f"Using environment: {env_name}")

    # Obtenir les informations de l'environnement (aprÃ¨s enregistrement)
    try:
        temp_env = gym.make(env_name)
        num_actions = temp_env.action_space.n
        # Assurez-vous que all_types et all_attributes sont correctement dÃ©finis
        num_event_types = len(all_types)
        num_colors = len(all_attributes.get("bg", [])) # Utiliser .get pour Ã©viter KeyError
        temp_env.close()
    except gym.error.Error as e:
        print(f"Error creating environment '{env_name}': {e}")
        print("Please ensure the environment is correctly registered and dependencies are installed.")
        return
    except NameError:
         print("Error: `all_types` or `all_attributes` not defined. Check `register_envs.py` or define defaults.")
         return


    env_info = {
        "num_actions": num_actions,
        "num_event_types": num_event_types,
        "num_colors": num_colors
    }
    print(f"Environment Info: {env_info}")

    # CrÃ©er l'agent
    agent = TransformerDQNAgent(env_info, CONFIG)

    # Optionnel: Charger un modÃ¨le prÃ©-entraÃ®nÃ© s'il existe
    if os.path.exists(CONFIG["save_path"]):
         agent.load_model(CONFIG["save_path"])
         # Peut-Ãªtre rÃ©initialiser epsilon si on veut continuer l'exploration
         # agent.epsilon = CONFIG["epsilon_start"] # DÃ©commenter si nÃ©cessaire


    # EntraÃ®ner l'agent
    print("\n--- Starting Training ---")
    train_rewards, train_durations = train_agent(agent, CONFIG)

    # Sauvegarder l'agent entraÃ®nÃ© final
    agent.save_model(CONFIG["save_path"] + "_final")
    print(f"Final model saved to {CONFIG['save_path']}_final")

    # Ã‰valuer l'agent entraÃ®nÃ©
    print("\n--- Starting Evaluation ---")
    eval_rewards = evaluate_agent(agent, CONFIG, num_episodes=100) # Ã‰valuer sur plus d'Ã©pisodes

    # Afficher les rÃ©sultats finaux
    final_avg_reward = np.mean(eval_rewards)
    print(f"\n--- Final Results ---")
    print(f"Average evaluation reward: {final_avg_reward:.2f}")

if __name__ == "__main__":
    main()
````

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import gym
import random
from collections import deque, namedtuple
import copy
import matplotlib.pyplot as plt
from torch.utils.tensorboard import SummaryWriter
import time
import math

# Configuration du modÃ¨le
class Config:
    def __init__(self):
        # ParamÃ¨tres du modÃ¨le
        self.event_dim = None  # Sera configurÃ© dynamiquement
        self.sequence_length = 20
        self.d_model = 128  # Dimension du modÃ¨le
        self.nhead = 8  # Nombre de tÃªtes d'attention
        self.num_encoder_layers = 4  # Nombre de couches d'encodeur
        self.num_decoder_layers = 0  # Pas besoin de dÃ©codeur pour cette tÃ¢che
        self.dim_feedforward = 512  # Dimension des couches feed-forward
        self.dropout = 0.1  # Taux de dropout
        
        # ParamÃ¨tres d'entraÃ®nement
        self.batch_size = 64
        self.lr = 1e-4
        self.gamma = 0.99  # Facteur de discount pour RL
        self.epsilon_start = 1.0
        self.epsilon_end = 0.01
        self.epsilon_decay = 10000
        self.target_update = 100  # FrÃ©quence de mise Ã  jour du rÃ©seau cible
        self.memory_size = 10000  # Taille du replay buffer
        
        # Autres paramÃ¨tres
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

config = Config()

# Classe pour traiter et encoder les Ã©vÃ©nements
class EventProcessor:
    def __init__(self, all_event_types, all_event_attributes, config):
        self.all_event_types = all_event_types
        self.all_event_attributes = all_event_attributes
        self.config = config
        
        # CrÃ©er des mappages pour l'encodage
        self.type_to_idx = {t: i for i, t in enumerate(all_event_types)}
        self.bg_to_idx = {c: i for i, c in enumerate(all_event_attributes['bg'])}
        self.fg_to_idx = {c: i for i, c in enumerate(all_event_attributes['fg'])}
        
        # Dimensions pour le one-hot encoding
        self.n_types = len(all_event_types)
        self.n_bg_colors = len(all_event_attributes['bg'])
        self.n_fg_colors = len(all_event_attributes['fg'])
        
        # Calculer la dimension totale du vecteur d'Ã©vÃ©nement
        # Type + bg + fg + temps (dÃ©but + durÃ©e)
        config.event_dim = self.n_types + self.n_bg_colors + self.n_fg_colors + 2
        print(f"Dimension d'encodage des Ã©vÃ©nements: {config.event_dim}")
    
    def encode_sequence(self, events):
        """
        Encode une sÃ©quence d'Ã©vÃ©nements
        """
        # VÃ©rifier qu'il y a des Ã©vÃ©nements Ã  encoder
        if not events:
            # Retourner une sÃ©quence vide si aucun Ã©vÃ©nement
            return np.zeros((self.config.sequence_length, self.config.event_dim), dtype=np.float32)
        
        # CrÃ©er un tableau de sortie avec la bonne forme
        result = np.zeros((self.config.sequence_length, self.config.event_dim), dtype=np.float32)
        
        # Encoder chaque Ã©vÃ©nement disponible
        for i, event in enumerate(events[-self.config.sequence_length:]):
            # Calculer la position correcte (Ã  la fin de la sÃ©quence)
            pos = max(0, self.config.sequence_length - len(events[-self.config.sequence_length:])) + i
            
            # Encoder l'Ã©vÃ©nement et vÃ©rifier sa dimension
            encoded_event = self.encode_event(event)
            
            # VÃ©rifier que la dimension correspond Ã  ce qui est attendu
            if len(encoded_event) != self.config.event_dim:
                print(f"ATTENTION: Dimension d'Ã©vÃ©nement incorrecte: {len(encoded_event)} vs {self.config.event_dim}")
                print(f"Ã‰vÃ©nement: {event}")
                # Adapter la taille si nÃ©cessaire (padding ou troncature)
                if len(encoded_event) < self.config.event_dim:
                    encoded_event = np.pad(encoded_event, (0, self.config.event_dim - len(encoded_event)))
                else:
                    encoded_event = encoded_event[:self.config.event_dim]
            
            result[pos] = encoded_event
        
        return result
    
    def encode_event(self, event):
        """
        Encode un Ã©vÃ©nement en vecteur de caractÃ©ristiques
        """
        # VÃ©rifier la structure de l'Ã©vÃ©nement
        if not isinstance(event, dict) or not all(key in event for key in ["symbol", "bg_color", "symbol_color", "start_time", "end_time"]):
            print(f"Structure d'Ã©vÃ©nement invalide: {event}")
            # Retourner un vecteur de zÃ©ros en cas d'Ã©vÃ©nement invalide
            return np.zeros(self.config.event_dim, dtype=np.float32)
        
        # One-hot encoding pour le type d'Ã©vÃ©nement
        type_vec = np.zeros(self.n_types, dtype=np.float32)
        if event["symbol"] in self.type_to_idx:
            type_vec[self.type_to_idx[event["symbol"]]] = 1.0
        
        # One-hot encoding pour les couleurs
        bg_vec = np.zeros(self.n_bg_colors, dtype=np.float32)
        if event["bg_color"] in self.bg_to_idx:
            bg_vec[self.bg_to_idx[event["bg_color"]]] = 1.0
        
        fg_vec = np.zeros(self.n_fg_colors, dtype=np.float32)
        if event["symbol_color"] in self.fg_to_idx:
            fg_vec[self.fg_to_idx[event["symbol_color"]]] = 1.0
        
        # Normalisation des temps (entre 0 et 1)
        start_time = np.float32(event["start_time"]) / 100.0
        duration = np.float32(event["end_time"] - event["start_time"]) / 20.0
        
        # ConcatÃ©ner le tout
        return np.concatenate([type_vec, bg_vec, fg_vec, [start_time, duration]])


# Architecture du Transformer pour RL
class TransformerModel(nn.Module):
    def __init__(self, config, n_actions):
        super(TransformerModel, self).__init__()
        self.config = config
        self.n_actions = n_actions
        
        # Couche d'embedding pour traiter les caractÃ©ristiques des Ã©vÃ©nements
        self.input_embedding = nn.Linear(config.event_dim, config.d_model)
        
        # Encodage positionnel
        self.pos_encoder = PositionalEncoding(config.d_model, config.dropout)
        
        # Couches de Transformer (encodeur uniquement)
        encoder_layers = nn.TransformerEncoderLayer(
            d_model=config.d_model,
            nhead=config.nhead,
            dim_feedforward=config.dim_feedforward,
            dropout=config.dropout,
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(
            encoder_layers,
            num_layers=config.num_encoder_layers
        )
        
        # Masque d'attention pour ignorer le padding
        self.register_buffer("src_mask", None)
        
        # Couche de sortie pour la prÃ©diction des Q-valeurs
        self.output = nn.Sequential(
            nn.Linear(config.d_model, config.dim_feedforward),
            nn.ReLU(),
            nn.Linear(config.dim_feedforward, n_actions)
        )
    
    def forward(self, src):
        # src: [batch_size, seq_len, event_dim]
        
        # CrÃ©ation du masque d'attention (pour ignorer le padding)
        # On considÃ¨re que les vecteurs d'Ã©vÃ©nements avec tous les Ã©lÃ©ments Ã  0 sont du padding
        src_key_padding_mask = (torch.sum(src, dim=2) == 0)
        
        # Embedding des caractÃ©ristiques d'entrÃ©e
        src = self.input_embedding(src)  # [batch_size, seq_len, d_model]
        
        # Ajout de l'encodage positionnel
        src = self.pos_encoder(src)
        
        # Passage dans le Transformer
        transformer_output = self.transformer_encoder(
            src, 
            mask=self.src_mask,
            src_key_padding_mask=src_key_padding_mask
        )  # [batch_size, seq_len, d_model]
        
        # On utilise la reprÃ©sentation de la derniÃ¨re position non-padding pour la dÃ©cision
        # Trouver le dernier indice non-padding pour chaque batch
        mask = ~src_key_padding_mask
        last_indices = torch.sum(mask, dim=1) - 1
        last_indices = torch.clamp(last_indices, min=0)
        
        # Extraire les caractÃ©ristiques Ã  ces indices
        batch_indices = torch.arange(transformer_output.size(0), device=transformer_output.device)
        output_features = transformer_output[batch_indices, last_indices]
        
        # PrÃ©diction des Q-values
        q_values = self.output(output_features)
        
        return q_values


# Encodage positionnel pour le Transformer
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, dropout=0.1, max_len=5000):
        super(PositionalEncoding, self).__init__()
        self.dropout = nn.Dropout(p=dropout)

        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(
            torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model)
        )
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)
        self.register_buffer("pe", pe)

    def forward(self, x):
        # x: [batch_size, seq_len, d_model]
        x = x + self.pe[:x.size(1), :].transpose(0, 1)
        return self.dropout(x)


# ImplÃ©mentation de l'agent DQN basÃ© sur Transformer
class TransformerDQNAgent:
    def __init__(self, event_processor, n_actions, config):
        self.event_processor = event_processor
        self.n_actions = n_actions
        self.config = config
        
        # RÃ©seaux policy et target
        self.policy_net = TransformerModel(config, n_actions).to(config.device)
        self.target_net = TransformerModel(config, n_actions).to(config.device)
        self.target_net.load_state_dict(self.policy_net.state_dict())
        self.target_net.eval()
        
        # Optimiseur
        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=config.lr)
        
        # MÃ©moire de replay
        self.memory = ReplayMemory(config.memory_size)
        
        # Compteur pour l'exploration
        self.steps_done = 0
        
        # Pour le tracking des mÃ©triques
        self.writer = SummaryWriter()
        self.episode_rewards = []
        self.training_losses = []
    
    def select_action(self, events):
        """
        SÃ©lectionne une action basÃ©e sur la politique epsilon-greedy
        """
        # Calculer epsilon pour l'exploration
        epsilon = self.config.epsilon_end + (self.config.epsilon_start - self.config.epsilon_end) * \
            math.exp(-1. * self.steps_done / self.config.epsilon_decay)
        self.steps_done += 1
        
        # Exploration
        if random.random() < epsilon:
            return random.randrange(self.n_actions)
        
        # Exploitation
        with torch.no_grad():
            # Encoder la sÃ©quence d'Ã©vÃ©nements
            state = self.event_processor.encode_sequence(events)
            state = torch.FloatTensor(state).unsqueeze(0).to(self.config.device)
            
            # Obtenir les Q-values du rÃ©seau policy
            q_values = self.policy_net(state)
            
            # SÃ©lectionner l'action avec la Q-value maximale
            return q_values.max(1)[1].item()
    
    def store_transition(self, state_events, action, next_state_events, reward, done):
        """
        Stocke une transition dans la mÃ©moire de replay
        """
        # Encoder les Ã©tats
        state = self.event_processor.encode_sequence(state_events)
        next_state = self.event_processor.encode_sequence(next_state_events)
        
        # Correction: S'assurer que les Ã©tats sont bien des tableaux numpy de type float32
        state = np.array(state, dtype=np.float32)
        next_state = np.array(next_state, dtype=np.float32)
        
        # Stocker la transition
        self.memory.push(
            torch.FloatTensor(state),
            torch.tensor([action]),
            torch.FloatTensor(next_state),
            torch.tensor([reward], dtype=torch.float),
            torch.tensor([done], dtype=torch.bool)
        )
    
    def optimize_model(self):
        """
        Effectue une Ã©tape d'optimisation sur le modÃ¨le
        """
        if len(self.memory) < self.config.batch_size:
            return
        
        # Ã‰chantillonner un batch de transitions
        transitions = self.memory.sample(self.config.batch_size)
        batch = Transition(*zip(*transitions))
        
        # CrÃ©er des tensors batch Ã  partir des transitions
        state_batch = torch.stack(batch.state).to(self.config.device)
        action_batch = torch.cat(batch.action).to(self.config.device)
        next_state_batch = torch.stack(batch.next_state).to(self.config.device)
        reward_batch = torch.cat(batch.reward).to(self.config.device)
        done_batch = torch.cat(batch.done).to(self.config.device)
        
        # VÃ©rifier et ajuster les dimensions de action_batch
        if action_batch.dim() == 1:
            action_batch = action_batch.unsqueeze(1)  # Ajouter une dimension pour obtenir [batch_size, 1]
        
        # Calcul des Q-values pour les actions prises
        q_values = self.policy_net(state_batch).gather(1, action_batch)
        
        # Calcul des Q-values futures (avec le rÃ©seau target)
        next_q_values = self.target_net(next_state_batch).max(1)[0].detach()
        
        # Calcul des Q-targets en utilisant l'Ã©quation de Bellman
        expected_q_values = reward_batch + (self.config.gamma * next_q_values * (~done_batch))
        
        # Calcul de la perte (erreur quadratique moyenne)
        loss = F.smooth_l1_loss(q_values, expected_q_values.unsqueeze(1))
        
        # Optimisation
        self.optimizer.zero_grad()
        loss.backward()
        
        # Gradient clipping pour stabilitÃ©
        for param in self.policy_net.parameters():
            param.grad.data.clamp_(-1, 1)
        
        self.optimizer.step()
        
        # Tracker la perte
        self.training_losses.append(loss.item())
        self.writer.add_scalar('Loss/train', loss.item(), self.steps_done)
        
        return loss.item()
    
    def update_target_network(self):
        """
        Met Ã  jour le rÃ©seau target avec les poids du rÃ©seau policy
        """
        self.target_net.load_state_dict(self.policy_net.state_dict())
    
    def save_model(self, path):
        """
        Sauvegarde le modÃ¨le
        """
        torch.save({
            'policy_net': self.policy_net.state_dict(),
            'target_net': self.target_net.state_dict(),
            'optimizer': self.optimizer.state_dict(),
            'steps_done': self.steps_done,
        }, path)
    
    def load_model(self, path):
        """
        Charge un modÃ¨le sauvegardÃ©
        """
        checkpoint = torch.load(path)
        self.policy_net.load_state_dict(checkpoint['policy_net'])
        self.target_net.load_state_dict(checkpoint['target_net'])
        self.optimizer.load_state_dict(checkpoint['optimizer'])
        self.steps_done = checkpoint['steps_done']


# Structure pour le replay buffer
Transition = namedtuple('Transition',
                        ('state', 'action', 'next_state', 'reward', 'done'))


# Classe ReplayMemory pour stocker les transitions
class ReplayMemory:
    def __init__(self, capacity):
        self.memory = deque(maxlen=capacity)
    
    def push(self, *args):
        """Sauvegarde une transition"""
        self.memory.append(Transition(*args))
    
    def sample(self, batch_size):
        """Ã‰chantillonne un batch de transitions"""
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)


# Fonction d'entraÃ®nement principale
def train_agent(env_name, agent, num_episodes=1000, max_steps_per_episode=100):
    """
    EntraÃ®ne l'agent DQN dans l'environnement
    """
    env = gym.make(env_name)
    
    all_rewards = []
    episode_durations = []
    
    for episode in range(num_episodes):
        # Initialiser l'environnement
        obs = env.reset()
        # Convertir l'observation en Ã©vÃ©nement
        current_event = event_from_obs_gym(obs, all_types, all_attributes)
        events_history = [current_event]
        
        episode_reward = 0
        
        # Boucle d'Ã©pisode
        for t in range(max_steps_per_episode):
            # SÃ©lectionner une action
            action = agent.select_action(events_history)
            
            # Convertir l'action pour l'environnement (peut nÃ©cessiter des ajustements)
            env_action = np.zeros(env.action_space.n)
            env_action[action] = 1
            
            # ExÃ©cuter l'action
            next_obs, reward, done, _ = env.step(env_action)
            
            # Convertir la nouvelle observation en Ã©vÃ©nement
            next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
            next_events_history = events_history + [next_event]
            
            # Limiter l'historique Ã  la longueur de sÃ©quence
            if len(next_events_history) > agent.config.sequence_length:
                next_events_history = next_events_history[-agent.config.sequence_length:]
            
            # Stocker la transition
            agent.store_transition(events_history, action, next_events_history, reward, done)
            
            # Passer Ã  l'Ã©tat suivant
            events_history = next_events_history
            episode_reward += reward
            
            # Optimiser le modÃ¨le
            loss = agent.optimize_model()
            
            # Mettre Ã  jour le rÃ©seau target pÃ©riodiquement
            if t % agent.config.target_update == 0:
                agent.update_target_network()
            
            if done:
                break
        
        # Fin de l'Ã©pisode
        all_rewards.append(episode_reward)
        episode_durations.append(t + 1)
        
        # Logging
        agent.writer.add_scalar('Reward/episode', episode_reward, episode)
        agent.writer.add_scalar('Duration/episode', t + 1, episode)
        
        # Afficher les progrÃ¨s
        if episode % 10 == 0:
            avg_reward = np.mean(all_rewards[-10:])
            print(f"Episode {episode}/{num_episodes}, Avg Reward: {avg_reward:.2f}, Epsilon: {agent.config.epsilon_end + (agent.config.epsilon_start - agent.config.epsilon_end) * math.exp(-1. * agent.steps_done / agent.config.epsilon_decay):.2f}")
            
            # Sauvegarder le modÃ¨le pÃ©riodiquement
            if episode % 100 == 0:
                agent.save_model(f"transformer_dqn_checkpoint_{episode}.pt")
    
    print("EntraÃ®nement terminÃ©!")
    agent.save_model("transformer_dqn_final.pt")
    env.close()
    
    return all_rewards, episode_durations


# Fonction pour Ã©valuer l'agent entraÃ®nÃ©
def evaluate_agent(env_name, agent, num_episodes=10):
    """
    Ã‰value les performances de l'agent
    """
    env = gym.make(env_name)
    rewards = []
    
    for episode in range(num_episodes):
        obs = env.reset()
        current_event = event_from_obs_gym(obs, all_types, all_attributes)
        events_history = [current_event]
        
        episode_reward = 0
        done = False
        
        while not done:
            # SÃ©lectionner la meilleure action (sans exploration)
            with torch.no_grad():
                state = agent.event_processor.encode_sequence(events_history)
                state = torch.FloatTensor(state).unsqueeze(0).to(agent.config.device)
                q_values = agent.policy_net(state)
                action = q_values.max(1)[1].item()
            
            # Convertir l'action pour l'environnement
            env_action = np.zeros(env.action_space.n)
            env_action[action] = 1
            
            # ExÃ©cuter l'action
            next_obs, reward, done, _ = env.step(env_action)
            
            # Mettre Ã  jour l'historique des Ã©vÃ©nements
            next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
            events_history.append(next_event)
            if len(events_history) > agent.config.sequence_length:
                events_history = events_history[-agent.config.sequence_length:]
            
            episode_reward += reward
        
        rewards.append(episode_reward)
        print(f"Ã‰pisode d'Ã©valuation {episode+1}/{num_episodes}, RÃ©compense: {episode_reward}")
    
    mean_reward = np.mean(rewards)
    print(f"RÃ©compense moyenne sur {num_episodes} Ã©pisodes: {mean_reward}")
    
    return rewards


# Fonction principale pour exÃ©cuter l'entraÃ®nement
def main():
    # DÃ©finir les types d'Ã©vÃ©nements et attributs
    all_event_types = ['A', 'B', 'C', 'D', 'E']
    all_event_attributes = {
        'fg': ['red', 'blue', 'green', 'orange', 'pink'],
        'bg': ['red', 'blue', 'green', 'orange', 'pink']
    }
    
    # Configuration initiale
    config = Config()
    
    # Initialiser le processeur d'Ã©vÃ©nements
    event_processor = EventProcessor(all_event_types, all_event_attributes, config)
    
    # VÃ©rifier qu'event_dim est correctement initialisÃ©
    print(f"Configuration event_dim: {config.event_dim}")
    
    # CrÃ©er l'environnement pour obtenir l'espace d'action
    env = gym.make("OpenTheChests-v2")
    n_actions = env.action_space.n
    env.close()
    
    # Initialiser l'agent
    agent = TransformerDQNAgent(event_processor, n_actions, config)
    
    # EntraÃ®ner l'agent
    rewards, durations = train_agent("OpenTheChests-v2", agent, num_episodes=500)
    
    # Ã‰valuer l'agent
    eval_rewards = evaluate_agent("OpenTheChests-v2", agent)
    
    # Visualiser les rÃ©sultats
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(rewards)
    plt.title('RÃ©compenses par Ã©pisode')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('RÃ©compense totale')
    
    plt.subplot(1, 2, 2)
    plt.plot(agent.training_losses)
    plt.title('Perte d\'entraÃ®nement')
    plt.xlabel('Ã‰tape d\'optimisation')
    plt.ylabel('Perte')
    
    plt.tight_layout()
    plt.savefig('training_results.png')
    plt.show()


if __name__ == "__main__":
    main()
````

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque
import math

class PositionalEncoding(nn.Module):
    """Encodage positionnel pour le Transformer"""
    def __init__(self, d_model, max_len=100):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]

class TransformerNetwork(nn.Module):
    """RÃ©seau basÃ© sur l'architecture Transformer pour le traitement de sÃ©quences"""
    def __init__(self, input_size, hidden_size, output_size, nhead=4, num_layers=2, dropout=0.1):
        super(TransformerNetwork, self).__init__()
        
        # Couche d'embedding linÃ©aire pour projeter les entrÃ©es dans l'espace cachÃ©
        self.input_embedding = nn.Linear(input_size, hidden_size)
        
        # Encodage positionnel
        self.pos_encoder = PositionalEncoding(hidden_size)
        
        # Couches d'encodeur Transformer
        encoder_layers = nn.TransformerEncoderLayer(d_model=hidden_size, nhead=nhead, 
                                                  dim_feedforward=hidden_size*4, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_layers)
        
        # Couche de sortie
        self.output_layer = nn.Linear(hidden_size, output_size)
        
        # Initialisation des poids
        self._init_weights()
        
    def _init_weights(self):
        """Initialisation des poids pour une meilleure convergence"""
        for p in self.parameters():
            if p.dim() > 1:
                nn.init.xavier_uniform_(p)
                
    def forward(self, src):
        # src shape: [batch_size, seq_len, input_size]
        
        # EntrÃ©e Ã  travers la couche d'embedding
        src = self.input_embedding(src)  # [batch_size, seq_len, hidden_size]
        
        # Ajout de l'encodage positionnel
        src = self.pos_encoder(src)
        
        # Transposer pour le format attendu par le transformer: [seq_len, batch_size, hidden_size]
        src = src.permute(1, 0, 2)
        
        # Passer Ã  travers l'encodeur transformer
        output = self.transformer_encoder(src)  # [seq_len, batch_size, hidden_size]
        
        # Utiliser seulement la sortie de la derniÃ¨re position
        output = output[-1]  # [batch_size, hidden_size]
        
        # Projection vers l'espace de sortie
        output = self.output_layer(output)  # [batch_size, output_size]
        
        return output

class TransformerAgent:
    """Agent utilisant un transformer pour traiter des sÃ©quences d'Ã©vÃ©nements complexes"""
    def __init__(self, state_size, action_size, hidden_size=64, sequence_length=16):
        self.state_size = state_size
        self.action_size = action_size
        self.hidden_size = hidden_size
        self.sequence_length = sequence_length
        
        # ParamÃ¨tres d'apprentissage
        self.gamma = 0.95  # facteur d'actualisation
        self.epsilon = 1.0  # taux d'exploration
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        
        # MÃ©moire de replay
        self.memory = deque(maxlen=5000)
        
        # Buffer de sÃ©quence pour stocker les Ã©vÃ©nements rÃ©cents
        self.sequence_buffer = []
        
        # ModÃ¨les de rÃ©seau
        self.model = TransformerNetwork(state_size, hidden_size, action_size)
        self.target_model = TransformerNetwork(state_size, hidden_size, action_size)
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.criterion = nn.MSELoss()
        
        # Synchronisation initiale des poids
        self.update_target_model()
    
    def update_target_model(self):
        """Copie les poids du modÃ¨le principal vers le modÃ¨le cible"""
        self.target_model.load_state_dict(self.model.state_dict())
    
    def encode_event(self, event):
        """
        Encode un Ã©vÃ©nement en vecteur pour l'entrÃ©e du rÃ©seau
        
        Args:
            event: dict contenant les informations sur l'Ã©vÃ©nement
                - symbol: type d'Ã©vÃ©nement (A, B, C, etc.)
                - bg_color: couleur d'arriÃ¨re-plan
                - symbol_color: couleur du symbole
                - start_time: temps de dÃ©but
                - end_time: temps de fin
                
        Returns:
            np.array: vecteur encodÃ© de l'Ã©vÃ©nement
        """
        # Encodage one-hot pour le symbole
        all_symbols = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
        symbol_vec = np.zeros(len(all_symbols))
        if event["symbol"] in all_symbols:
            symbol_vec[all_symbols.index(event["symbol"])] = 1
        
        # Encodage one-hot pour les couleurs
        all_colors = ['red', 'blue', 'green', 'orange', 'pink', 'yellow', 'purple', 'black']
        bg_vec = np.zeros(len(all_colors))
        if event["bg_color"] in all_colors:
            bg_vec[all_colors.index(event["bg_color"])] = 1
            
        fg_vec = np.zeros(len(all_colors))
        if event["symbol_color"] in all_colors:
            fg_vec[all_colors.index(event["symbol_color"])] = 1
        
        # Temps normalisÃ©s et relations temporelles
        start_time = event["start_time"] / 100  # Normalisation arbitraire
        end_time = event["end_time"] / 100
        duration = end_time - start_time
        
        # ConcatÃ©ner tous les vecteurs d'encodage
        return np.concatenate([symbol_vec, bg_vec, fg_vec, [start_time, end_time, duration]])
    
    def remember(self, state, action, reward, next_state, done):
        """Stocke l'expÃ©rience dans la mÃ©moire de replay"""
        self.memory.append((state, action, reward, next_state, done))
    
    def act(self, event):
        """Choisit une action basÃ©e sur l'Ã©vÃ©nement actuel et l'historique"""
        # Encode l'Ã©vÃ©nement et ajoute-le au buffer de sÃ©quence
        event_vector = self.encode_event(event)
        self.sequence_buffer.append(event_vector)
        
        # Garde seulement la longueur de sÃ©quence dÃ©finie
        if len(self.sequence_buffer) > self.sequence_length:
            self.sequence_buffer.pop(0)
        
        # Padde la sÃ©quence si nÃ©cessaire
        padded_sequence = self.sequence_buffer.copy()
        while len(padded_sequence) < self.sequence_length:
            padded_sequence.insert(0, np.zeros_like(event_vector))
        
        # Convertit en tensor pour le rÃ©seau
        state = torch.FloatTensor(padded_sequence).unsqueeze(0)  # Ajoute dimension batch
        
        # Epsilon-greedy pour l'exploration
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        
        # PrÃ©diction du modÃ¨le
        self.model.eval()
        with torch.no_grad():
            q_values = self.model(state)
        self.model.train()
        
        return torch.argmax(q_values).item()
    
    def replay(self, batch_size):
        """EntraÃ®ne le modÃ¨le sur un mini-batch d'expÃ©riences"""
        if len(self.memory) < batch_size:
            return
        
        # Ã‰chantillonne un mini-batch de la mÃ©moire
        minibatch = random.sample(self.memory, batch_size)
        
        losses = []
        for state, action, reward, next_state, done in minibatch:
            state = torch.FloatTensor(state).unsqueeze(0)
            next_state = torch.FloatTensor(next_state).unsqueeze(0)
            
            # Calcul de la cible Q
            self.model.eval()
            with torch.no_grad():
                current_q = self.model(state)
                target_q = current_q.clone()
                
                if done:
                    target_q[0][action] = reward
                else:
                    # Double DQN: utiliser le modÃ¨le principal pour sÃ©lectionner l'action,
                    # et le modÃ¨le cible pour Ã©valuer sa valeur
                    next_q = self.model(next_state)
                    next_action = torch.argmax(next_q).item()
                    
                    next_target_q = self.target_model(next_state)
                    target_q[0][action] = reward + self.gamma * next_target_q[0][next_action]
            
            # EntraÃ®ne le modÃ¨le
            self.model.train()
            self.optimizer.zero_grad()
            current_q = self.model(state)
            loss = self.criterion(current_q, target_q)
            loss.backward()
            self.optimizer.step()
            
            losses.append(loss.item())
        
        # DÃ©croissance du taux d'exploration
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
        
        # Mise Ã  jour pÃ©riodique du modÃ¨le cible
        self.update_target_model()
        
        return np.mean(losses)
    
    def load(self, name):
        """Charge les poids du modÃ¨le"""
        self.model.load_state_dict(torch.load(name))
        self.update_target_model()
    
    def save(self, name):
        """Sauvegarde les poids du modÃ¨le"""
        torch.save(self.model.state_dict(), name)


# Exemple d'utilisation 
def train_transformer_agent(env, episodes=1000, batch_size=32):
    """
    EntraÃ®ne l'agent Transformer sur l'environnement OpenTheChestsGym
    
    Args:
        env: environnement OpenTheChestsGym
        episodes: nombre d'Ã©pisodes d'entraÃ®nement
        batch_size: taille du mini-batch pour l'apprentissage
        
    Returns:
        agent: agent Transformer entraÃ®nÃ©
        scores: historique des scores
    """
    # DÃ©terminer les dimensions d'Ã©tat et d'action
    state_size = 27  # 8 symboles + 8 couleurs bg + 8 couleurs fg + 3 temps
    action_size = env.action_space.n
    
    # Initialiser l'agent
    agent = TransformerAgent(state_size, action_size)
    
    # Variables de suivi
    scores = []
    
    for e in range(episodes):
        # RÃ©initialiser l'environnement et l'Ã©tat
        obs = env.reset()
        event = event_from_obs_gym(obs, all_types, all_attributes)
        
        # RÃ©initialiser le buffer de sÃ©quence
        agent.sequence_buffer = []
        
        total_reward = 0
        done = False
        
        while not done:
            # Construire l'Ã©tat actuel
            current_state = agent.sequence_buffer.copy()
            while len(current_state) < agent.sequence_length:
                current_state.insert(0, np.zeros(state_size))
            
            # Choisir une action
            action_idx = agent.act(event)
            
            # AprÃ¨s examen du code d'erreur, il semble que nous devions utiliser l'action directement comme un entier
            # et non comme une liste
            
            # ExÃ©cuter l'action
            action = [action_idx] * env.action_space.n
            next_obs, reward, done, info = env.step(action)
            total_reward += reward
            
            # Construire le prochain Ã©tat
            if "context" in next_obs and next_obs["context"] is not None:
                next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
                
                # Encoder le prochain Ã©vÃ©nement
                next_event_vector = agent.encode_event(next_event)
                
                # Mettre Ã  jour le buffer de sÃ©quence pour le prochain Ã©tat
                next_state = agent.sequence_buffer.copy() + [next_event_vector]
                if len(next_state) > agent.sequence_length:
                    next_state = next_state[-agent.sequence_length:]
                while len(next_state) < agent.sequence_length:
                    next_state.insert(0, np.zeros(state_size))
            else:
                next_state = current_state
                next_event = event
            
            # Stocker l'expÃ©rience dans la mÃ©moire
            agent.remember(current_state, action_idx, reward, next_state, done)
            
            # Passer Ã  l'Ã©tat suivant
            event = next_event
            
            # EntraÃ®ner le modÃ¨le
            if len(agent.memory) > batch_size:
                loss = agent.replay(batch_size)
        
        scores.append(total_reward)
        
        # Afficher les progrÃ¨s
        if e % 10 == 0:
            print(f"Episode: {e}/{episodes}, Score: {total_reward}, Epsilon: {agent.epsilon:.2f}")
            
        # Sauvegarder le modÃ¨le pÃ©riodiquement
        if e % 100 == 0:
            agent.save(f"transformer_agent_checkpoint_{e}.pth")
    
    # Sauvegarder le modÃ¨le final
    agent.save("transformer_agent_final.pth")
    
    return agent, scores


# Fonction d'Ã©valuation
def evaluate_agent(env, agent, num_episodes=10):
    """
    Ã‰value les performances d'un agent entraÃ®nÃ©
    
    Args:
        env: environnement OpenTheChestsGym
        agent: agent entraÃ®nÃ©
        num_episodes: nombre d'Ã©pisodes d'Ã©valuation
        
    Returns:
        mean_score: score moyen sur les Ã©pisodes d'Ã©valuation
    """
    scores = []
    
    for e in range(num_episodes):
        obs = env.reset()
        event = event_from_obs_gym(obs, all_types, all_attributes)
        
        agent.sequence_buffer = []
        total_reward = 0
        done = False
        
        while not done:
            # DÃ©sactiver l'exploration pendant l'Ã©valuation
            epsilon_backup = agent.epsilon
            agent.epsilon = 0
            
            action_idx = agent.act(event)
            # Convertir l'indice d'action en format attendu par l'environnement
            action = [0] * env.action_space.n
            action[action_idx] = 1
            
            next_obs, reward, done, _ = env.step(action)
            total_reward += reward
            
            # Restaurer epsilon
            agent.epsilon = epsilon_backup
            
            if "context" in next_obs and next_obs["context"] is not None:
                event = event_from_obs_gym(next_obs, all_types, all_attributes) 
        scores.append(total_reward)
        print(f"Episode d'Ã©valuation {e+1}: Score = {total_reward}")
    
    mean_score = np.mean(scores)
    print(f"Score moyen sur {num_episodes} Ã©pisodes: {mean_score}")
    return mean_score

# Fonctions d'analyse et de visualisation
def visualize_training_progress(scores, window_size=10):
    """
    Visualise la progression de l'entraÃ®nement
    
    Args:
        scores: historique des scores d'entraÃ®nement
        window_size: taille de la fenÃªtre pour le calcul de la moyenne mobile
    """
    import matplotlib.pyplot as plt
    
    # Calculer la moyenne mobile
    moving_avg = np.convolve(scores, np.ones(window_size)/window_size, mode='valid')
    
    plt.figure(figsize=(12, 6))
    plt.plot(scores, alpha=0.3, color='blue', label='Scores individuels')
    plt.plot(np.arange(window_size-1, len(scores)), moving_avg, color='red', label=f'Moyenne mobile ({window_size})')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('Score')
    plt.title('Progression de l\'entraÃ®nement')
    plt.legend()
    plt.grid(True)
    plt.show()


def visualize_event_sequence_and_actions(env, agent, max_steps=50):
    """
    Visualise une sÃ©quence d'Ã©vÃ©nements et les actions prises par l'agent
    
    Args:
        env: environnement OpenTheChestsGym
        agent: agent entraÃ®nÃ©
        max_steps: nombre maximum d'Ã©tapes Ã  visualiser
    """
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    
    # RÃ©initialiser l'environnement et l'agent
    obs = env.reset()
    agent.sequence_buffer = []
    
    # Collecter les Ã©vÃ©nements et actions
    events = []
    actions = []
    rewards = []
    
    event = event_from_obs_gym(obs, all_types, all_attributes)
    
    # DÃ©sactiver l'exploration
    epsilon_backup = agent.epsilon
    agent.epsilon = 0
    
    for step in range(max_steps):
        action_idx = agent.act(event)
        actions.append(action_idx)
        
        # Convertir l'indice d'action en format attendu par l'environnement
        action = [0] * env.action_space.n
        action[action_idx] = 1
        
        next_obs, reward, done, _ = env.step(action)
        rewards.append(reward)
        
        if "context" in next_obs and next_obs["context"] is not None:
            event = event_from_obs_gym(next_obs, all_types, all_attributes)
            events.append(event)
        
        if done:
            break
    
    # Restaurer epsilon
    agent.epsilon = epsilon_backup
    
    # CrÃ©er la visualisation
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8), gridspec_kw={'height_ratios': [3, 1]})
    
    # Tracer la sÃ©quence d'Ã©vÃ©nements
    max_time = max([e["end_time"] for e in events])
    height = 1
    last_event_end_times = []
    
    for event in events:
        event_name = event["symbol"]
        start = event["start_time"]
        end = event["end_time"]
        color = event["bg_color"]
        text_color = event["symbol_color"]
        
        # Trouver la ligne correcte pour l'Ã©vÃ©nement
        line = 0
        while line < len(last_event_end_times):
            if start >= last_event_end_times[line]:
                break
            line += 1
        if line == len(last_event_end_times):
            last_event_end_times.append(end)
        else:
            last_event_end_times[line] = end
        
        # Dessiner l'Ã©vÃ©nement
        y_pos = line * (height + 0.5)
        rect = patches.Rectangle((start, y_pos), max(end - start, 0.1), height, color=color, alpha=0.7)
        ax1.add_patch(rect)
        
        # Annoter l'Ã©vÃ©nement
        ax1.text(start + (end - start) / 2, y_pos + height / 2, event_name,
                horizontalalignment='center', verticalalignment='center',
                color=text_color, fontsize=12, fontweight='bold')
    
    # Configurer le premier axe
    ax1.set_xlim(0, max_time)
    ax1.set_ylim(0, (len(last_event_end_times) + 1) * (height + 0.5))
    ax1.set_xlabel('Temps')
    ax1.set_ylabel('Ã‰vÃ©nements')
    ax1.set_title('SÃ©quence d\'Ã©vÃ©nements observÃ©s')
    
    # Tracer les actions et rÃ©compenses
    step_times = np.arange(len(actions))
    ax2.bar(step_times, [1] * len(actions), color='lightgray', alpha=0.7)
    
    for i, (action, reward) in enumerate(zip(actions, rewards)):
        color = 'green' if reward > 0 else 'red' if reward < 0 else 'blue'
        ax2.text(i, 0.5, f'A{action}\nR:{reward}', 
                horizontalalignment='center', verticalalignment='center',
                color=color, fontsize=10, fontweight='bold')
    
    # Configurer le second axe
    ax2.set_xlim(-0.5, len(actions) - 0.5)
    ax2.set_ylim(0, 1)
    ax2.set_xlabel('Ã‰tape')
    ax2.set_ylabel('Action')
    ax2.set_title('Actions prises par l\'agent')
    ax2.set_yticks([])
    
    plt.tight_layout()
    plt.show()


# Point d'entrÃ©e principal pour l'exÃ©cution du code
if __name__ == "__main__":
    # Ce bloc s'exÃ©cute uniquement si le script est lancÃ© directement
    
    # Exemple d'utilisation avec l'environnement de niveau 2 (difficile)
    
    env = gym.make("OpenTheChests-v2")
    print("Environnement chargÃ© avec succÃ¨s.")
    
    # EntraÃ®ner l'agent (commentez pour charger un modÃ¨le existant)
    agent, training_scores = train_transformer_agent(env, episodes=500, batch_size=64)
    
    # Visualiser la progression de l'entraÃ®nement
    visualize_training_progress(training_scores)
    
    # Ã‰valuer l'agent
    evaluate_agent(env, agent, num_episodes=5)
    
    # Visualiser une sÃ©quence d'Ã©vÃ©nements et les actions de l'agent
    visualize_event_sequence_and_actions(env, agent)
    
    # Fermer l'environnement
    env.close()
````

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque
import math
import gym  # Assurez-vous d'avoir gym installÃ©

class PositionalEncoding(nn.Module):
    """Encodage positionnel pour le Transformer"""
    def __init__(self, d_model, max_len=100):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]

class TransformerNetwork(nn.Module):
    """RÃ©seau basÃ© sur l'architecture Transformer pour le traitement de sÃ©quences"""
    def __init__(self, input_size, hidden_size, output_size, nhead=4, num_layers=2, dropout=0.1):
        super(TransformerNetwork, self).__init__()
        
        # Couche d'embedding linÃ©aire pour projeter les entrÃ©es dans l'espace cachÃ©
        self.input_embedding = nn.Linear(input_size, hidden_size)
        
        # Encodage positionnel
        self.pos_encoder = PositionalEncoding(hidden_size)
        
        # Couches d'encodeur Transformer
        encoder_layers = nn.TransformerEncoderLayer(d_model=hidden_size, nhead=nhead, 
                                                  dim_feedforward=hidden_size*4, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_layers)
        
        # Couche de sortie
        self.output_layer = nn.Linear(hidden_size, output_size)
        
        # Initialisation des poids
        self._init_weights()
        
    def _init_weights(self):
        """Initialisation des poids pour une meilleure convergence"""
        for p in self.parameters():
            if p.dim() > 1:
                nn.init.xavier_uniform_(p)
                
    def forward(self, src):
        # src shape: [batch_size, seq_len, input_size]
        src = self.input_embedding(src)  # [batch_size, seq_len, hidden_size]
        src = self.pos_encoder(src)
        src = src.permute(1, 0, 2)  # Format attendu: [seq_len, batch_size, hidden_size]
        output = self.transformer_encoder(src)  # [seq_len, batch_size, hidden_size]
        output = output[-1]  # On utilise uniquement la derniÃ¨re sortie: [batch_size, hidden_size]
        output = self.output_layer(output)  # Projection vers l'espace d'action
        return output

class TransformerAgent:
    """Agent utilisant un transformer pour traiter des sÃ©quences d'Ã©vÃ©nements complexes"""
    def __init__(self, state_size, action_size, hidden_size=64, sequence_length=16):
        self.state_size = state_size
        self.action_size = action_size
        self.hidden_size = hidden_size
        self.sequence_length = sequence_length
        
        # ParamÃ¨tres d'apprentissage
        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        
        # MÃ©moire de replay
        self.memory = deque(maxlen=5000)
        
        # Buffer de sÃ©quence pour stocker les Ã©vÃ©nements rÃ©cents
        self.sequence_buffer = []
        
        # ModÃ¨les de rÃ©seau
        self.model = TransformerNetwork(state_size, hidden_size, action_size)
        self.target_model = TransformerNetwork(state_size, hidden_size, action_size)
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.criterion = nn.MSELoss()
        
        # Synchronisation initiale des poids
        self.update_target_model()
    
    def update_target_model(self):
        """Copie les poids du modÃ¨le principal vers le modÃ¨le cible"""
        self.target_model.load_state_dict(self.model.state_dict())
    
    def encode_event(self, event):
        """
        Encode un Ã©vÃ©nement en vecteur pour l'entrÃ©e du rÃ©seau
        
        Args:
            event: dict contenant les informations sur l'Ã©vÃ©nement
                - symbol: type d'Ã©vÃ©nement (A, B, C, etc.)
                - bg_color: couleur d'arriÃ¨re-plan
                - symbol_color: couleur du symbole
                - start_time: temps de dÃ©but
                - end_time: temps de fin
                
        Returns:
            np.array: vecteur encodÃ© de l'Ã©vÃ©nement
        """
        # Encodage one-hot pour le symbole
        all_symbols = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
        symbol_vec = np.zeros(len(all_symbols))
        if event["symbol"] in all_symbols:
            symbol_vec[all_symbols.index(event["symbol"])] = 1
        
        # Encodage one-hot pour les couleurs
        all_colors = ['red', 'blue', 'green', 'orange', 'pink', 'yellow', 'purple', 'black']
        bg_vec = np.zeros(len(all_colors))
        if event["bg_color"] in all_colors:
            bg_vec[all_colors.index(event["bg_color"])] = 1
            
        fg_vec = np.zeros(len(all_colors))
        if event["symbol_color"] in all_colors:
            fg_vec[all_colors.index(event["symbol_color"])] = 1
        
        # Temps normalisÃ©s
        start_time = event["start_time"] / 100
        end_time = event["end_time"] / 100
        duration = end_time - start_time
        
        return np.concatenate([symbol_vec, bg_vec, fg_vec, [start_time, end_time, duration]])
    
    def remember(self, state, action, reward, next_state, done):
        """Stocke l'expÃ©rience dans la mÃ©moire de replay"""
        self.memory.append((state, action, reward, next_state, done))
    
    def act(self, event):
        """Choisit une action basÃ©e sur l'Ã©vÃ©nement actuel et l'historique"""
        # Encodage et ajout au buffer de sÃ©quence
        event_vector = self.encode_event(event)
        self.sequence_buffer.append(event_vector)
        
        # Garde la sÃ©quence Ã  la bonne longueur
        if len(self.sequence_buffer) > self.sequence_length:
            self.sequence_buffer.pop(0)
        
        # Padding si nÃ©cessaire
        padded_sequence = self.sequence_buffer.copy()
        while len(padded_sequence) < self.sequence_length:
            padded_sequence.insert(0, np.zeros_like(event_vector))
        
        # Passage en tensor
        state = torch.FloatTensor(padded_sequence).unsqueeze(0)
        
        # Epsilon-greedy
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        
        self.model.eval()
        with torch.no_grad():
            q_values = self.model(state)
        self.model.train()
        
        return torch.argmax(q_values).item()
    
    def replay(self, batch_size):
        """EntraÃ®ne le modÃ¨le sur un mini-batch d'expÃ©riences"""
        if len(self.memory) < batch_size:
            return
        
        minibatch = random.sample(self.memory, batch_size)
        losses = []
        for state, action, reward, next_state, done in minibatch:
            state = torch.FloatTensor(state).unsqueeze(0)
            next_state = torch.FloatTensor(next_state).unsqueeze(0)
            
            self.model.eval()
            with torch.no_grad():
                current_q = self.model(state)
                target_q = current_q.clone()
                
                if done:
                    target_q[0][action] = reward
                else:
                    # Double DQN
                    next_q = self.model(next_state)
                    next_action = torch.argmax(next_q).item()
                    next_target_q = self.target_model(next_state)
                    target_q[0][action] = reward + self.gamma * next_target_q[0][next_action]
            
            self.model.train()
            self.optimizer.zero_grad()
            current_q = self.model(state)
            loss = self.criterion(current_q, target_q)
            loss.backward()
            self.optimizer.step()
            
            losses.append(loss.item())
        
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
        
        self.update_target_model()
        return np.mean(losses)
    
    def load(self, name):
        """Charge les poids du modÃ¨le"""
        self.model.load_state_dict(torch.load(name))
        self.update_target_model()
    
    def save(self, name):
        """Sauvegarde les poids du modÃ¨le"""
        torch.save(self.model.state_dict(), name)


# Exemple d'utilisation 
def train_transformer_agent(env, episodes=1000, batch_size=32):
    """
    EntraÃ®ne l'agent Transformer sur l'environnement OpenTheChestsGym
    
    Args:
        env: environnement OpenTheChestsGym
        episodes: nombre d'Ã©pisodes d'entraÃ®nement
        batch_size: taille du mini-batch pour l'apprentissage
        
    Returns:
        agent: agent Transformer entraÃ®nÃ©
        scores: historique des scores
    """
    state_size = 27  # 8 symboles + 8 couleurs bg + 8 couleurs fg + 3 valeurs de temps
    action_size = env.action_space.n
    
    agent = TransformerAgent(state_size, action_size)
    
    scores = []
    
    for e in range(episodes):
        obs = env.reset()
        event = event_from_obs_gym(obs, all_types, all_attributes)
        
        # RÃ©initialise le buffer de sÃ©quence
        agent.sequence_buffer = []
        total_reward = 0
        done = False
        
        while not done:
            current_state = agent.sequence_buffer.copy()
            while len(current_state) < agent.sequence_length:
                current_state.insert(0, np.zeros(state_size))
            
            # Choix de l'action : l'agent retourne directement un indice
            action_idx = agent.act(event)
            
            # Utiliser l'action sous forme d'entier pour env.step()
            action = action_idx  
            next_obs, reward, done, info = env.step(action)
            total_reward += reward
            
            if "context" in next_obs and next_obs["context"] is not None:
                next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
                next_event_vector = agent.encode_event(next_event)
                next_state = agent.sequence_buffer.copy() + [next_event_vector]
                if len(next_state) > agent.sequence_length:
                    next_state = next_state[-agent.sequence_length:]
                while len(next_state) < agent.sequence_length:
                    next_state.insert(0, np.zeros(state_size))
            else:
                next_state = current_state
                next_event = event
            
            agent.remember(current_state, action_idx, reward, next_state, done)
            event = next_event
            
            if len(agent.memory) > batch_size:
                loss = agent.replay(batch_size)
        
        scores.append(total_reward)
        if e % 10 == 0:
            print(f"Episode: {e}/{episodes}, Score: {total_reward}, Epsilon: {agent.epsilon:.2f}")
        if e % 100 == 0:
            agent.save(f"transformer_agent_checkpoint_{e}.pth")
    
    agent.save("transformer_agent_final.pth")
    return agent, scores


def evaluate_agent(env, agent, num_episodes=10):
    """
    Ã‰value les performances d'un agent entraÃ®nÃ©
    
    Args:
        env: environnement OpenTheChestsGym
        agent: agent entraÃ®nÃ©
        num_episodes: nombre d'Ã©pisodes d'Ã©valuation
        
    Returns:
        mean_score: score moyen sur les Ã©pisodes d'Ã©valuation
    """
    scores = []
    
    for e in range(num_episodes):
        obs = env.reset()
        event = event_from_obs_gym(obs, all_types, all_attributes)
        agent.sequence_buffer = []
        total_reward = 0
        done = False
        
        while not done:
            epsilon_backup = agent.epsilon
            agent.epsilon = 0  # DÃ©sactive l'exploration pendant l'Ã©valuation
            
            action_idx = agent.act(event)
            # Utiliser l'indice d'action directement
            action = action_idx
            next_obs, reward, done, _ = env.step(action)
            total_reward += reward
            
            agent.epsilon = epsilon_backup  # Restaure epsilon
            
            if "context" in next_obs and next_obs["context"] is not None:
                event = event_from_obs_gym(next_obs, all_types, all_attributes)
        scores.append(total_reward)
        print(f"Episode d'Ã©valuation {e+1}: Score = {total_reward}")
    
    mean_score = np.mean(scores)
    print(f"Score moyen sur {num_episodes} Ã©pisodes: {mean_score}")
    return mean_score


def visualize_training_progress(scores, window_size=10):
    """
    Visualise la progression de l'entraÃ®nement
    """
    import matplotlib.pyplot as plt
    moving_avg = np.convolve(scores, np.ones(window_size)/window_size, mode='valid')
    
    plt.figure(figsize=(12, 6))
    plt.plot(scores, alpha=0.3, color='blue', label='Scores individuels')
    plt.plot(np.arange(window_size-1, len(scores)), moving_avg, color='red', label=f'Moyenne mobile ({window_size})')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('Score')
    plt.title('Progression de l\'entraÃ®nement')
    plt.legend()
    plt.grid(True)
    plt.show()


def visualize_event_sequence_and_actions(env, agent, max_steps=50):
    """
    Visualise une sÃ©quence d'Ã©vÃ©nements et les actions prises par l'agent
    """
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    
    obs = env.reset()
    agent.sequence_buffer = []
    events = []
    actions = []
    rewards = []
    
    event = event_from_obs_gym(obs, all_types, all_attributes)
    
    epsilon_backup = agent.epsilon
    agent.epsilon = 0
    
    for step in range(max_steps):
        action_idx = agent.act(event)
        actions.append(action_idx)
        
        # Utiliser l'action directement comme indice
        action = action_idx
        next_obs, reward, done, _ = env.step(action)
        rewards.append(reward)
        
        if "context" in next_obs and next_obs["context"] is not None:
            event = event_from_obs_gym(next_obs, all_types, all_attributes)
            events.append(event)
        
        if done:
            break
    
    agent.epsilon = epsilon_backup
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8), gridspec_kw={'height_ratios': [3, 1]})
    
    max_time = max([e["end_time"] for e in events]) if events else 1
    height = 1
    last_event_end_times = []
    
    for event in events:
        event_name = event["symbol"]
        start = event["start_time"]
        end = event["end_time"]
        color = event["bg_color"]
        text_color = event["symbol_color"]
        
        line = 0
        while line < len(last_event_end_times):
            if start >= last_event_end_times[line]:
                break
            line += 1
        if line == len(last_event_end_times):
            last_event_end_times.append(end)
        else:
            last_event_end_times[line] = end
        
        y_pos = line * (height + 0.5)
        rect = patches.Rectangle((start, y_pos), max(end - start, 0.1), height, color=color, alpha=0.7)
        ax1.add_patch(rect)
        ax1.text(start + (end - start) / 2, y_pos + height / 2, event_name,
                horizontalalignment='center', verticalalignment='center',
                color=text_color, fontsize=12, fontweight='bold')
    
    ax1.set_xlim(0, max_time)
    ax1.set_ylim(0, (len(last_event_end_times) + 1) * (height + 0.5))
    ax1.set_xlabel('Temps')
    ax1.set_ylabel('Ã‰vÃ©nements')
    ax1.set_title('SÃ©quence d\'Ã©vÃ©nements observÃ©s')
    
    step_times = np.arange(len(actions))
    ax2.bar(step_times, [1] * len(actions), color='lightgray', alpha=0.7)
    
    for i, (action, reward) in enumerate(zip(actions, rewards)):
        color = 'green' if reward > 0 else 'red' if reward < 0 else 'blue'
        ax2.text(i, 0.5, f'A{action}\nR:{reward}', 
                horizontalalignment='center', verticalalignment='center',
                color=color, fontsize=10, fontweight='bold')
    
    ax2.set_xlim(-0.5, len(actions) - 0.5)
    ax2.set_ylim(0, 1)
    ax2.set_xlabel('Ã‰tape')
    ax2.set_ylabel('Action')
    ax2.set_title('Actions prises par l\'agent')
    ax2.set_yticks([])
    
    plt.tight_layout()
    plt.show()


# Point d'entrÃ©e principal
if __name__ == "__main__":
    # Chargement de l'environnement
    env = gym.make("OpenTheChests-v2")
    print("Environnement chargÃ© avec succÃ¨s.")
    
    # EntraÃ®nement de l'agent
    agent, training_scores = train_transformer_agent(env, episodes=500, batch_size=64)
    visualize_training_progress(training_scores)
    evaluate_agent(env, agent, num_episodes=5)
    visualize_event_sequence_and_actions(env, agent)
    env.close()
````

````python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque
import math

# Configuration du device (CPU ou GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Utilisation de l'appareil: {device}")

class PositionalEncoding(nn.Module):
    """Encodage positionnel pour le Transformer"""
    def __init__(self, d_model, max_len=100):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :].to(x.device)

class TransformerNetwork(nn.Module):
    """RÃ©seau basÃ© sur l'architecture Transformer pour le traitement de sÃ©quences"""
    def __init__(self, input_size, hidden_size, output_size, nhead=4, num_layers=2, dropout=0.1):
        super(TransformerNetwork, self).__init__()
        
        # Couche d'embedding linÃ©aire pour projeter les entrÃ©es dans l'espace cachÃ©
        self.input_embedding = nn.Linear(input_size, hidden_size)
        
        # Encodage positionnel
        self.pos_encoder = PositionalEncoding(hidden_size)
        
        # Couches d'encodeur Transformer
        encoder_layers = nn.TransformerEncoderLayer(d_model=hidden_size, nhead=nhead, 
                                                  dim_feedforward=hidden_size*4, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_layers)
        
        # Couche de sortie
        self.output_layer = nn.Linear(hidden_size, output_size)
        
        # Initialisation des poids
        self._init_weights()
        
    def _init_weights(self):
        """Initialisation des poids pour une meilleure convergence"""
        for p in self.parameters():
            if p.dim() > 1:
                nn.init.xavier_uniform_(p)
                
    def forward(self, src):
        # src shape: [batch_size, seq_len, input_size]
        
        # EntrÃ©e Ã  travers la couche d'embedding
        src = self.input_embedding(src)  # [batch_size, seq_len, hidden_size]
        
        # Ajout de l'encodage positionnel
        src = self.pos_encoder(src)
        
        # Transposer pour le format attendu par le transformer: [seq_len, batch_size, hidden_size]
        src = src.permute(1, 0, 2)
        
        # Passer Ã  travers l'encodeur transformer
        output = self.transformer_encoder(src)  # [seq_len, batch_size, hidden_size]
        
        # Utiliser seulement la sortie de la derniÃ¨re position
        output = output[-1]  # [batch_size, hidden_size]
        
        # Projection vers l'espace de sortie
        output = self.output_layer(output)  # [batch_size, output_size]
        
        return output

class TransformerAgent:
    """Agent utilisant un transformer pour traiter des sÃ©quences d'Ã©vÃ©nements complexes"""
    def __init__(self, state_size, action_size, hidden_size=64, sequence_length=16):
        self.state_size = state_size
        self.action_size = action_size
        self.hidden_size = hidden_size
        self.sequence_length = sequence_length
        self.device = device  # Utiliser le device global
        
        # ParamÃ¨tres d'apprentissage
        self.gamma = 0.95  # facteur d'actualisation
        self.epsilon = 1.0  # taux d'exploration
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        
        # MÃ©moire de replay
        self.memory = deque(maxlen=5000)
        
        # Buffer de sÃ©quence pour stocker les Ã©vÃ©nements rÃ©cents
        self.sequence_buffer = []
        
        # ModÃ¨les de rÃ©seau
        self.model = TransformerNetwork(state_size, hidden_size, action_size).to(self.device)
        self.target_model = TransformerNetwork(state_size, hidden_size, action_size).to(self.device)
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.criterion = nn.MSELoss()
        
        # Synchronisation initiale des poids
        self.update_target_model()
    
    def update_target_model(self):
        """Copie les poids du modÃ¨le principal vers le modÃ¨le cible"""
        self.target_model.load_state_dict(self.model.state_dict())
    
    def encode_event(self, event):
        """
        Encode un Ã©vÃ©nement en vecteur pour l'entrÃ©e du rÃ©seau
        
        Args:
            event: dict contenant les informations sur l'Ã©vÃ©nement
                - symbol: type d'Ã©vÃ©nement (A, B, C, etc.)
                - bg_color: couleur d'arriÃ¨re-plan
                - symbol_color: couleur du symbole
                - start_time: temps de dÃ©but
                - end_time: temps de fin
                
        Returns:
            np.array: vecteur encodÃ© de l'Ã©vÃ©nement
        """
        # Encodage one-hot pour le symbole
        all_symbols = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
        symbol_vec = np.zeros(len(all_symbols))
        if event["symbol"] in all_symbols:
            symbol_vec[all_symbols.index(event["symbol"])] = 1
        
        # Encodage one-hot pour les couleurs
        all_colors = ['red', 'blue', 'green', 'orange', 'pink', 'yellow', 'purple', 'black']
        bg_vec = np.zeros(len(all_colors))
        if event["bg_color"] in all_colors:
            bg_vec[all_colors.index(event["bg_color"])] = 1
            
        fg_vec = np.zeros(len(all_colors))
        if event["symbol_color"] in all_colors:
            fg_vec[all_colors.index(event["symbol_color"])] = 1
        
        # Temps normalisÃ©s et relations temporelles
        start_time = event["start_time"] / 100  # Normalisation arbitraire
        end_time = event["end_time"] / 100
        duration = end_time - start_time
        
        # ConcatÃ©ner tous les vecteurs d'encodage
        return np.concatenate([symbol_vec, bg_vec, fg_vec, [start_time, end_time, duration]])
    
    def remember(self, state, action, reward, next_state, done):
        """Stocke l'expÃ©rience dans la mÃ©moire de replay"""
        self.memory.append((state, action, reward, next_state, done))
    
    def act(self, event):
        """Choisit une action basÃ©e sur l'Ã©vÃ©nement actuel et l'historique"""
        # Encode l'Ã©vÃ©nement et ajoute-le au buffer de sÃ©quence
        event_vector = self.encode_event(event)
        self.sequence_buffer.append(event_vector)
        
        # Garde seulement la longueur de sÃ©quence dÃ©finie
        if len(self.sequence_buffer) > self.sequence_length:
            self.sequence_buffer.pop(0)
        
        # Padde la sÃ©quence si nÃ©cessaire
        padded_sequence = self.sequence_buffer.copy()
        while len(padded_sequence) < self.sequence_length:
            padded_sequence.insert(0, np.zeros_like(event_vector))
        
        # Convertit en tensor pour le rÃ©seau et dÃ©place sur le device appropriÃ©
        state = torch.FloatTensor(padded_sequence).unsqueeze(0).to(self.device)  # Ajoute dimension batch
        
        # Epsilon-greedy pour l'exploration
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        
        # PrÃ©diction du modÃ¨le
        self.model.eval()
        with torch.no_grad():
            q_values = self.model(state)
        self.model.train()
        
        return torch.argmax(q_values).item()
    
    def replay(self, batch_size):
        """EntraÃ®ne le modÃ¨le sur un mini-batch d'expÃ©riences"""
        if len(self.memory) < batch_size:
            return
        
        # Ã‰chantillonne un mini-batch de la mÃ©moire
        minibatch = random.sample(self.memory, batch_size)
        
        losses = []
        for state, action, reward, next_state, done in minibatch:
            state = torch.FloatTensor(state).unsqueeze(0).to(self.device)
            next_state = torch.FloatTensor(next_state).unsqueeze(0).to(self.device)
            
            # Calcul de la cible Q
            self.model.eval()
            with torch.no_grad():
                current_q = self.model(state)
                target_q = current_q.clone()
                
                if done:
                    target_q[0][action] = reward
                else:
                    # Double DQN: utiliser le modÃ¨le principal pour sÃ©lectionner l'action,
                    # et le modÃ¨le cible pour Ã©valuer sa valeur
                    next_q = self.model(next_state)
                    next_action = torch.argmax(next_q).item()
                    
                    next_target_q = self.target_model(next_state)
                    target_q[0][action] = reward + self.gamma * next_target_q[0][next_action]
            
            # EntraÃ®ne le modÃ¨le
            self.model.train()
            self.optimizer.zero_grad()
            current_q = self.model(state)
            loss = self.criterion(current_q, target_q)
            loss.backward()
            self.optimizer.step()
            
            losses.append(loss.item())
        
        # DÃ©croissance du taux d'exploration
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
        
        # Mise Ã  jour pÃ©riodique du modÃ¨le cible
        self.update_target_model()
        
        return np.mean(losses)
    
    def load(self, name):
        """Charge les poids du modÃ¨le"""
        # Charge les poids sur le device appropriÃ©
        self.model.load_state_dict(torch.load(name, map_location=self.device))
        self.update_target_model()
    
    def save(self, name):
        """Sauvegarde les poids du modÃ¨le"""
        torch.save(self.model.state_dict(), name)


# Exemple d'utilisation 
def train_transformer_agent(env, episodes=1000, batch_size=32):
    """
    EntraÃ®ne l'agent Transformer sur l'environnement OpenTheChestsGym
    
    Args:
        env: environnement OpenTheChestsGym
        episodes: nombre d'Ã©pisodes d'entraÃ®nement
        batch_size: taille du mini-batch pour l'apprentissage
        
    Returns:
        agent: agent Transformer entraÃ®nÃ©
        scores: historique des scores
    """
    # DÃ©terminer les dimensions d'Ã©tat et d'action
    state_size = 27  # 8 symboles + 8 couleurs bg + 8 couleurs fg + 3 temps
    action_size = env.action_space.n
    
    # Initialiser l'agent
    agent = TransformerAgent(state_size, action_size)
    
    # Variables de suivi
    scores = []
    
    # Utiliser tqdm pour afficher une barre de progression si disponible
    try:
        from tqdm import tqdm
        episode_iterator = tqdm(range(episodes))
    except ImportError:
        episode_iterator = range(episodes)
    
    for e in episode_iterator:
        # RÃ©initialiser l'environnement et l'Ã©tat
        obs = env.reset()
        event = event_from_obs_gym(obs, all_types, all_attributes)
        
        # RÃ©initialiser le buffer de sÃ©quence
        agent.sequence_buffer = []
        
        total_reward = 0
        done = False
        
        while not done:
            # Construire l'Ã©tat actuel
            current_state = agent.sequence_buffer.copy()
            while len(current_state) < agent.sequence_length:
                current_state.insert(0, np.zeros(state_size))
            
            # Choisir une action
            action_idx = agent.act(event)
            
            # ExÃ©cuter l'action
            action = [action_idx] * env.action_space.n
            next_obs, reward, done, info = env.step(action)
            total_reward += reward
            
            # Construire le prochain Ã©tat
            if "context" in next_obs and next_obs["context"] is not None:
                next_event = event_from_obs_gym(next_obs, all_types, all_attributes)
                
                # Encoder le prochain Ã©vÃ©nement
                next_event_vector = agent.encode_event(next_event)
                
                # Mettre Ã  jour le buffer de sÃ©quence pour le prochain Ã©tat
                next_state = agent.sequence_buffer.copy() + [next_event_vector]
                if len(next_state) > agent.sequence_length:
                    next_state = next_state[-agent.sequence_length:]
                while len(next_state) < agent.sequence_length:
                    next_state.insert(0, np.zeros(state_size))
            else:
                next_state = current_state
                next_event = event
            
            # Stocker l'expÃ©rience dans la mÃ©moire
            agent.remember(current_state, action_idx, reward, next_state, done)
            
            # Passer Ã  l'Ã©tat suivant
            event = next_event
            
            # EntraÃ®ner le modÃ¨le
            if len(agent.memory) > batch_size:
                loss = agent.replay(batch_size)
        
        scores.append(total_reward)
        
        # Afficher les progrÃ¨s
        if e % 10 == 0:
            print(f"Episode: {e}/{episodes}, Score: {total_reward}, Epsilon: {agent.epsilon:.2f}")
            
        # Sauvegarder le modÃ¨le pÃ©riodiquement
        if e % 100 == 0:
            agent.save(f"transformer_agent_checkpoint_{e}.pth")
    
    # Sauvegarder le modÃ¨le final
    agent.save("transformer_agent_final.pth")
    
    return agent, scores


# Fonction d'Ã©valuation
def evaluate_agent(env, agent, num_episodes=10):
    """
    Ã‰value les performances d'un agent entraÃ®nÃ©
    
    Args:
        env: environnement OpenTheChestsGym
        agent: agent entraÃ®nÃ©
        num_episodes: nombre d'Ã©pisodes d'Ã©valuation
        
    Returns:
        mean_score: score moyen sur les Ã©pisodes d'Ã©valuation
    """
    scores = []
    
    for e in range(num_episodes):
        obs = env.reset()
        event = event_from_obs_gym(obs, all_types, all_attributes)
        
        agent.sequence_buffer = []
        total_reward = 0
        done = False
        
        while not done:
            # DÃ©sactiver l'exploration pendant l'Ã©valuation
            epsilon_backup = agent.epsilon
            agent.epsilon = 0
            
            action_idx = agent.act(event)
            # Convertir l'indice d'action en format attendu par l'environnement
            action = [0] * env.action_space.n
            action[action_idx] = 1
            
            next_obs, reward, done, _ = env.step(action)
            total_reward += reward
            
            # Restaurer epsilon
            agent.epsilon = epsilon_backup
            
            if "context" in next_obs and next_obs["context"] is not None:
                event = event_from_obs_gym(next_obs, all_types, all_attributes) 
        scores.append(total_reward)
        print(f"Episode d'Ã©valuation {e+1}: Score = {total_reward}")
    
    mean_score = np.mean(scores)
    print(f"Score moyen sur {num_episodes} Ã©pisodes: {mean_score}")
    return mean_score

# Fonctions d'analyse et de visualisation
def visualize_training_progress(scores, window_size=10):
    """
    Visualise la progression de l'entraÃ®nement
    
    Args:
        scores: historique des scores d'entraÃ®nement
        window_size: taille de la fenÃªtre pour le calcul de la moyenne mobile
    """
    import matplotlib.pyplot as plt
    
    # Calculer la moyenne mobile
    moving_avg = np.convolve(scores, np.ones(window_size)/window_size, mode='valid')
    
    plt.figure(figsize=(12, 6))
    plt.plot(scores, alpha=0.3, color='blue', label='Scores individuels')
    plt.plot(np.arange(window_size-1, len(scores)), moving_avg, color='red', label=f'Moyenne mobile ({window_size})')
    plt.xlabel('Ã‰pisode')
    plt.ylabel('Score')
    plt.title('Progression de l\'entraÃ®nement')
    plt.legend()
    plt.grid(True)
    plt.savefig('training_progress.png')  # Sauvegarde l'image
    plt.show()


def visualize_event_sequence_and_actions(env, agent, max_steps=50):
    """
    Visualise une sÃ©quence d'Ã©vÃ©nements et les actions prises par l'agent
    
    Args:
        env: environnement OpenTheChestsGym
        agent: agent entraÃ®nÃ©
        max_steps: nombre maximum d'Ã©tapes Ã  visualiser
    """
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    
    # RÃ©initialiser l'environnement et l'agent
    obs = env.reset()
    agent.sequence_buffer = []
    
    # Collecter les Ã©vÃ©nements et actions
    events = []
    actions = []
    rewards = []
    
    event = event_from_obs_gym(obs, all_types, all_attributes)
    
    # DÃ©sactiver l'exploration
    epsilon_backup = agent.epsilon
    agent.epsilon = 0
    
    for step in range(max_steps):
        action_idx = agent.act(event)
        actions.append(action_idx)
        
        # Convertir l'indice d'action en format attendu par l'environnement
        action = [0] * env.action_space.n
        action[action_idx] = 1
        
        next_obs, reward, done, _ = env.step(action)
        rewards.append(reward)
        
        if "context" in next_obs and next_obs["context"] is not None:
            event = event_from_obs_gym(next_obs, all_types, all_attributes)
            events.append(event)
        
        if done:
            break
    
    # Restaurer epsilon
    agent.epsilon = epsilon_backup
    
    # CrÃ©er la visualisation
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8), gridspec_kw={'height_ratios': [3, 1]})
    
    # Tracer la sÃ©quence d'Ã©vÃ©nements
    max_time = max([e["end_time"] for e in events]) if events else 10
    height = 1
    last_event_end_times = []
    
    for event in events:
        event_name = event["symbol"]
        start = event["start_time"]
        end = event["end_time"]
        color = event["bg_color"]
        text_color = event["symbol_color"]
        
        # Trouver la ligne correcte pour l'Ã©vÃ©nement
        line = 0
        while line < len(last_event_end_times):
            if start >= last_event_end_times[line]:
                break
            line += 1
        if line == len(last_event_end_times):
            last_event_end_times.append(end)
        else:
            last_event_end_times[line] = end
        
        # Dessiner l'Ã©vÃ©nement
        y_pos = line * (height + 0.5)
        rect = patches.Rectangle((start, y_pos), max(end - start, 0.1), height, color=color, alpha=0.7)
        ax1.add_patch(rect)
        
        # Annoter l'Ã©vÃ©nement
        ax1.text(start + (end - start) / 2, y_pos + height / 2, event_name,
                horizontalalignment='center', verticalalignment='center',
                color=text_color, fontsize=12, fontweight='bold')
    
    # Configurer le premier axe
    ax1.set_xlim(0, max_time)
    ax1.set_ylim(0, (len(last_event_end_times) + 1) * (height + 0.5))
    ax1.set_xlabel('Temps')
    ax1.set_ylabel('Ã‰vÃ©nements')
    ax1.set_title('SÃ©quence d\'Ã©vÃ©nements observÃ©s')
    
    # Tracer les actions et rÃ©compenses
    step_times = np.arange(len(actions))
    ax2.bar(step_times, [1] * len(actions), color='lightgray', alpha=0.7)
    
    for i, (action, reward) in enumerate(zip(actions, rewards)):
        color = 'green' if reward > 0 else 'red' if reward < 0 else 'blue'
        ax2.text(i, 0.5, f'A{action}\nR:{reward}', 
                horizontalalignment='center', verticalalignment='center',
                color=color, fontsize=10, fontweight='bold')
    
    # Configurer le second axe
    ax2.set_xlim(-0.5, len(actions) - 0.5)
    ax2.set_ylim(0, 1)
    ax2.set_xlabel('Ã‰tape')
    ax2.set_ylabel('Action')
    ax2.set_title('Actions prises par l\'agent')
    ax2.set_yticks([])
    
    plt.tight_layout()
    plt.savefig('event_sequence.png')  # Sauvegarde l'image
    plt.show()


# Point d'entrÃ©e principal pour l'exÃ©cution du code
if __name__ == "__main__":
    # Importation des modules manquants en haut du script
    import gym
    
    # DÃ©finir les fonctions manquantes dans le code original
    def event_from_obs_gym(obs, all_types, all_attributes):
        # Cette fonction doit Ãªtre dÃ©finie selon le format de votre environnement
        # Voici une implÃ©mentation basique supposÃ©e
        if "context" not in obs or obs["context"] is None:
            # Valeurs par dÃ©faut si contexte non disponible
            return {
                "symbol": "A",
                "bg_color": "blue",
                "symbol_color": "white",
                "start_time": 0,
                "end_time": 1
            }
        
        context = obs["context"]
        # Supposer que context contient les attributs nÃ©cessaires
        return {
            "symbol": context.get("symbol", "A"),
            "bg_color": context.get("background_color", "blue"),
            "symbol_color": context.get("symbol_color", "white"),
            "start_time": context.get("start_time", 0),
            "end_time": context.get("end_time", 1)
        }
    
    # DÃ©finition des variables globales utilisÃ©es dans les fonctions
    all_types = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
    all_attributes = ['red', 'blue', 'green', 'orange', 'pink', 'yellow', 'purple', 'black']
    
    # Configuration CUDA optimisÃ©e si disponible
    if torch.cuda.is_available():
        torch.backends.cudnn.benchmark = True
        print(f"GPU disponible: {torch.cuda.get_device_name(0)}")
        print(f"Nombre de GPUs: {torch.cuda.device_count()}")
    
    # Ce bloc s'exÃ©cute uniquement si le script est lancÃ© directement
    try:
        # Exemple d'utilisation avec l'environnement de niveau 2 (difficile)
        env = gym.make("OpenTheChests-v2")
        print("Environnement chargÃ© avec succÃ¨s.")
        
        # EntraÃ®ner l'agent (commentez pour charger un modÃ¨le existant)
        agent, training_scores = train_transformer_agent(env, episodes=500, batch_size=128)
        
        # Visualiser la progression de l'entraÃ®nement
        visualize_training_progress(training_scores)
        
        # Ã‰valuer l'agent
        evaluate_agent(env, agent, num_episodes=5)
        
        # Visualiser une sÃ©quence d'Ã©vÃ©nements et les actions de l'agent
        visualize_event_sequence_and_actions(env, agent)
        
        # Fermer l'environnement
        env.close()
    except Exception as e:
        print(f"Erreur lors de l'exÃ©cution: {e}")
````

## 3. Combining the Two Environments

### **Objective**
The goal of this step is to **integrate robotic control** with the **Open the Chests** environment, creating a **dual-agent system** where one agent focuses on **event recognition**, and the other handles **robotic movement and task execution**.

### **Two-Agent System**
The system consists of **two specialized agents** that work together to achieve the objective:

#### **Agent 1: Event Recognition Agent**
- **Role:** Identifies when it is time to open a chest based on the observed event stream.
- **Task:** Recognizes **patterns in event sequences** and determines the correct **moment for action**.
- **Challenge:** Must handle **asynchronous event streams** and recognize **sequential dependencies**.

#### **Agent 2: Robotic Control Agent**
- **Role:** Controls the **robotic arm**, moving it to the correct chest and performing the action of opening it.
- **Task:** Receives signals from **Agent 1**, executes movement, and performs the physical **interaction with the chest**.
- **Challenge:** Accurately **execute motor control** and properly **coordinate** with **Agent 1**.

### **Goal**
The **dual-agent system** must function **cooperatively**, where:
- **Agent 1** handles **activity recognition** and determines when a chest should be opened.
- **Agent 2** receives the **signal from Agent 1**, moves the **robotic arm**, and executes the **chest-opening action**.

### **Suggested Approach**
To effectively **solve this problem**, the following **strategy** is recommended:

1. **Event Recognition (Agent 1)**
   - **Process the event stream** using a suitable pattern recognition method.
   - Possible techniques include:
     - **Temporal Windows** â€“ Track events over a time window to detect relevant sequences.
     - **Transformers / LSTMs** â€“ Use deep learning models to analyze complex event dependencies.
   - Once a **chest-opening pattern** is detected, **send a signal** to Agent 2.

2. **Robotic Control (Agent 2)**
   - **Receive the signal** from Agent 1.
   - Move the **robotic arm** to the identified chest.
   - Perform the **physical interaction**.
   - Confirm the **completion of the task** to the Open the Chests (OtC) environment.

3. **Cycle Reset & Continuation**
   - After the chest is opened:
     - **Agent 1 resumes event observation** to detect the next relevant sequence.
     - The **robot arm may reset** to an initial position before processing the next task.
   - The cycle repeats, allowing for **continuous operation**.
